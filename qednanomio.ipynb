```python
import time
import torch

def evaluate_inference(pipe, prompt, gen_config):
    # Convertiamo il prompt in formato chat se è una stringa
    if isinstance(prompt, str):
        formatted_prompt = [{"role": "user", "content": prompt}]
    else:
        formatted_prompt = prompt

    start_time = time.time()

    # Esecuzione dell'inferenza
    result = pipe(formatted_prompt, generation_config=gen_config)

    end_time = time.time()
    duration = end_time - start_time

    # Estrazione del testo generato (formato chat)
    generated_text = result[0]['generated_text'][-1]['content']

    # Calcolo approssimativo dei token
    num_tokens = len(generated_text.split())

    print(f"--- Performance Report ---")
    print(f"Tempo totale: {duration:.2f} secondi")
    print(f"Token generati (stima parole): {num_tokens}")
    print(f"Throughput: {num_tokens / duration:.2f} token/s")
    print(f"\nRisposta: {generated_text}")

# Esempio di test
evaluate_inference(pipe, "Spiega brevemente la fotosintesi.", generation_config)
```

### Metriche Chiave per Modelli Linguistici

1. **Perplessità (Perplexity):** Misura quanto il modello è 'sorpreso' da un dataset di test. Più bassa è, migliore è la capacità predittiva.
2. **Latenza:** Il tempo necessario per generare il primo token o l'intera sequenza.
3. **Benchmark Standard:** Utilizzare strumenti come `EleutherAI/lm-evaluation-harness` per testare il modello su task di ragionamento logico, matematico o medico.

```python
from transformers import pipeline, GenerationConfig

pipe = pipeline("text-generation", model="lm-provers/QED-Nano")
messages = [
    {"role": "user", "content": "What is the Pythagorean theorem?"},
]

generation_config = GenerationConfig(
    max_new_tokens=256
)

pipe(messages, generation_config=generation_config)
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("lm-provers/QED-Nano")
model = AutoModelForCausalLM.from_pretrained("lm-provers/QED-Nano")
messages = [
    {"role": "user", "content": "Give a practical example or analogy explaining how GFR decreases during a downstream obstruction, using the concept of hydrostatic pressure."},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

# Using 256 tokens for a clear practical example
outputs = model.generate(**inputs, max_new_tokens=256)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
```

```python
import torch
import gc
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline, GenerationConfig, BitsAndBytesConfig

def clear_gpu():
    if 'pipe' in locals(): del pipe
    if 'model' in locals(): del model
    gc.collect()
    torch.cuda.empty_cache()

clear_gpu()

model_id = "lm-provers/QED-Nano"
bnb_config = BitsAndBytesConfig(load_in_4bit=True, bnb_4bit_compute_dtype=torch.float16, bnb_4bit_quant_type="nf4")

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id, quantization_config=bnb_config, device_map="auto")
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

generation_config = GenerationConfig(max_new_tokens=800, temperature=0.1)

messages = [
    {"role": "user", "content": "Paziente: uomo 71 anni, dolore retrosternale oppressivo da sforzo (3 rampe scale) da 3 settimane, regredisce con riposo. Sintomi stabili. Opzioni: Angina microbiotica, Angina instabile, Scompenso cardiaco, Angina domestica, Angina di Prinzmetal. Qual è la diagnosi corretta tra queste? Spiega il motivo per cui scegli quella opzione e perché escludi le altre, considerando che il quadro sembra di angina stabile ma non è tra le opzioni."}
]

response = pipe(messages, generation_config=generation_config)
print(response[0]['generated_text'][-1]['content'])
```
