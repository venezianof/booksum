````python
simulation_script = """
import time
import sys
import json

def run_simulation(dosage):
    start_time = time.time()
    time.sleep(0.1)
    efficacy = 100 * (1 - (dosage - 0.5)**2)
    toxicity = 50 * (dosage**2)
    latency = time.time() - start_time
    return {
        "efficacy_score": round(max(0, efficacy), 2),
        "toxicity_level": round(toxicity, 2),
        "latency": round(latency, 2)
    }

if __name__ == "__main__":
    dosage = 0.3
    results = run_simulation(dosage)
    print(json.dumps(results))
    if results['toxicity_level'] > 15:
        sys.exit(1)
"""
with open('simulate_treatment.py', 'w') as f:
    f.write(simulation_script)
print('Script simulate_treatment.py ripristinato.')
````

### Avvio Loop Autoresearch: Ottimizzazione Dosaggio
Configuriamo lo script che gestirà le iterazioni di ottimizzazione su `simulate_treatment.py`.

````python
import json
import subprocess
import os
import re

def run_iteration(dosage):
    """Esegue una singola iterazione di test con un dosaggio specifico."""
    try:
        with open('simulate_treatment.py', 'r') as f:
            lines = f.readlines()

        with open('simulate_treatment.py', 'w') as f:
            for line in lines:
                if 'dosage =' in line and 'def' not in line:
                    f.write(f'    dosage = {dosage}\n')
                else:
                    f.write(line)

        output = subprocess.check_output(['python3', 'simulate_treatment.py'], stderr=subprocess.STDOUT).decode()
        match = re.search(r'\{.*\}', output, re.DOTALL)
        return json.loads(match.group()) if match else None
    except Exception as e:
        return None

# --- Logica di Ottimizzazione Dinamica ---
print("--- Avvio Ottimizzazione Dinamica del Dosaggio ---")

current_dosage = 0.1
step = 0.05
best_result = None
max_iterations = 10
safety_limit = 15.0

for i in range(max_iterations):
    res = run_iteration(round(current_dosage, 2))

    if res:
        print(f"Iterazione {i+1}: Dosaggio {current_dosage:.2f} -> Efficacia: {res['efficacy_score']}, Tossicità: {res['toxicity_level']}")

        if res['toxicity_level'] >= safety_limit:
            print("⚠️ Soglia di sicurezza raggiunta. Interruzione per prevenire tossicità.")
            break

        if best_result is None or res['efficacy_score'] > best_result['efficacy_score']:
            best_result = res
            best_result['dosage'] = current_dosage
            current_dosage += step
        else:
            # Se l'efficacia cala, riduciamo lo step e proviamo a raffinare
            step /= 2
            current_dosage -= step
    else:
        print(f"Iterazione {i+1}: Fallimento critico (probabile tossicità eccessiva).")
        break

if best_result:
    print(f"\nRisultato Ottimale: Dosaggio {best_result['dosage']:.2f} con Efficacia {best_result['efficacy_score']}")
````

### Il Ruolo dei Sintomi: Valutazione ABE (GOLD 2024)

Le linee guida hanno semplificato lo schema ABCD nel nuovo modello **ABE**. I sintomi determinano se un paziente appartiene alla colonna di sinistra (pochi sintomi) o di destra (molti sintomi).

*   **Pochi Sintomi:** CAT < 10 oppure mMRC 0-1.
*   **Molti Sintomi:** CAT ≥ 10 oppure mMRC ≥ 2.

La combinazione con il rischio di esacerbazioni definisce il gruppo:
- **Gruppo A:** Pochi sintomi, basso rischio esacerbazioni.
- **Gruppo B:** Molti sintomi, basso rischio esacerbazioni.
- **Gruppo E:** Alto rischio esacerbazioni (indipendentemente dai sintomi).

### Focus: Criteri GOLD 2024 - Gruppo E (Exacerbations)

Il **Gruppo E** è stato introdotto per unificare i precedenti gruppi C e D, semplificando la scelta terapeutica iniziale basata sul rischio clinico.

#### 1. Criteri di Inclusione (Basta uno dei seguenti):
- **≥ 2 esacerbazioni moderate** nell'ultimo anno (trattate con antibiotici e/o corticosteroidi sistemici).

#### 2. Indipendenza dai Sintomi:
- A differenza dei gruppi A e B, la classificazione nel Gruppo E avviene **indipendentemente dal punteggio mMRC o CAT**.
- Anche un paziente con pochi sintomi (es. CAT < 10) viene inserito nel Gruppo E se soddisfa i criteri di esacerbazione sopra elencati.

#### 3. Implicazioni Terapeutiche:
- L'obiettivo principale è la riduzione del rischio futuro.
- La terapia iniziale raccomandata è solitamente la **LABA+LAMA**.
- Si considera la **Tripla Terapia (LABA+LAMA+ICS)** se gli eosinofili ematici sono ≥ 300 cellule/αL o in caso di riacutizzazioni persistenti.

### Linee Guida GINA per la Diagnosi di Asma nei Bambini

La diagnosi di asma si basa sulla combinazione di **sintomi respiratori caratteristici** e l'evidenza di una **limitazione variabile del flusso aereo**.

#### 1. Bambini di età ≤ 5 anni (Diagnosi Clinica)
In questa fascia d'età, la spirometria è difficile da eseguire in modo affidabile. La diagnosi è prevalentemente clinica:
- **Sintomi**: Dispnea, respiro sibilante (wheezing), tosse e senso di costrizione toracica.
- **Pattern**: I sintomi peggiorano di notte o al risveglio, sono scatenati da esercizio, riso, pianto o aria fredda.
- **Prova Terapeutica**: Un miglioramento clinico significativo durante 2-3 mesi di terapia con ICS (corticosteroidi inalatori) al bisogno o regolarmente, e peggioramento dopo la sospensione, supporta fortemente la diagnosi.

#### 2. Bambini di età > 5 anni (Diagnosi Obiettiva)
Oltre ai sintomi clinici, è necessaria la conferma oggettiva della variazione del flusso aereo.

| Test | Criterio GINA per la Diagnosi |
| :--- | :--- |
| **Spirometria (Reversibilit&agrave;)** | Aumento del **FEV1 > 12%** rispetto al basale dopo 200-400 μg di salbutamolo. |
| **Variabilit&agrave; del PEF** | Variabilit&agrave; giornaliera media > 13% (calcolata su 2 settimane). |
| **Test da Sforzo** | Caduta del **FEV1 > 12%** (o del PEF > 15%) dal valore pre-esercizio. |
| **Test di Provocazione (Metacolina)** | Caduta del FEV1 ≥ 20% con dosi standard (test molto sensibile per escludere l'asma). |
| **Variabilit&agrave; Inter-visita** | Variazione del FEV1 > 12% tra visite diverse (non legata a infezioni respiratorie). |

#### Punti Chiave per la Diagnosi Differenziale
GINA sottolinea l'importanza di escludere alternative comuni nei bambini:
- Infezioni respiratorie ricorrenti.
- Reflusso gastroesofageo.
- Aspirazione di corpo estraneo (esordio improvviso).
- Discinesia ciliare primaria o Fibrosi Cistica.
- Malformazioni congenite delle vie aeree.

### Analisi Caso Clinico: Presbiopia e Correzione Ottica

**Definizione**: La presbiopia è un errore refrattivo dovuto alla perdita fisiologica dell'elasticità del cristallino e della funzione del muscolo ciliare.

#### Opzioni di Management:
- **Lente Convessa (Positiva)**: È la scelta corretta. Funge da "addizione" di potere convergente, permettendo di focalizzare sulla retina le immagini degli oggetti posti a distanza ravvicinata.
- **Lente Concava (Negativa)**: Diverge i raggi luminosi. Si utilizza per correggere la miopia.
- **Lente Prismatica**: Devia il fascio luminoso senza cambiare la messa a fuoco; indicata per disturbi della motilità oculare (diplopia).
- **Bifocali**: Possono essere usati se il paziente ha anche un difetto da lontano, ma il quesito specifica che la visione da lontano è buona.

**Risposta al quesito**: Lente convessa per visione da vicino.

### Farmacogenetica: Carbamazepina e Reazioni Avverse Gravi

**Contesto**: La carbamazepina è un anticonvulsivante di prima linea, ma il suo utilizzo è limitato dal rischio di ipersensibilità cutanea.

#### Il Marker Genetico: HLA-B*15:02
- **Associazione**: Esiste una correlazione estremamente forte tra la presenza dell'allele **HLA-B*15:02** e l'insorgenza della Sindrome di Stevens-Johnson (SJS) o della Necrolisi Epidermica Tossica (TEN).
- **Popolazioni target**: Lo screening è obbligatorio o fortemente raccomandato nelle popolazioni del sud-est asiatico (Han cinesi, tailandesi, malesi, indiani), dove l'allele è più frequente.
- **Significato Clinico**: Se il test è positivo, la carbamazepina deve essere evitata a favore di altri antiepilettici (a meno che i benefici non superino chiaramente i rischi).

#### Altri Test Citati (Diagnosi Differenziale):
- **HLA-B*58:01**: Screening per rischio SJS/TEN da **Allopurinolo**.
- **VKORC1 / CYP2C9**: Ottimizzazione del dosaggio di **Warfarin**.
- **CYP2D6**: Metabolismo di antidepressivi, oppioidi e tamoxifene.

### 🔍 Agente di Ricerca Medline Avanzato
Questo agente implementa una logica di ricerca profonda includendo:
- **Sorting**: Pertinenza o Data di pubblicazione.
- **Date Filtering**: Ricerca limitata agli ultimi N anni.
- **MeSH Terms**: Estrazione delle parole chiave strutturate per comprendere il focus dell'articolo.

````python
!pip install -q biopython
import pandas as pd
from Bio import Entrez
import time

class MedlineAgent:
    def __init__(self, email="your_email@example.com"):
        Entrez.email = email

    def deep_search(self, query, max_results=10, sort='relevance', years_back=None):
        """
        Esegue una ricerca approfondita su PubMed.
        sort: 'relevance', 'pub_date'
        """
        print(f"🚀 Avvio ricerca per: {query}...")

        # Calcolo filtro data se richiesto
        date_filter = ""
        if years_back:
            from datetime import datetime
            current_year = datetime.now().year
            date_filter = f" AND ({current_year - years_back}:{current_year}[pdat])"
            query += date_filter

        try:
            # 1. Ricerca ID
            handle = Entrez.esearch(db="pubmed", term=query, retmax=max_results, sort=sort)
            record = Entrez.read(handle)
            handle.close()
            id_list = record["IdList"]

            if not id_list:
                return "Nessun risultato trovato."

            # 2. Recupero Dettagli XML
            handle = Entrez.efetch(db="pubmed", id=",".join(id_list), retmode="xml")
            records = Entrez.read(handle)
            handle.close()

            parsed_results = []
            for article in records['PubmedArticle']:
                medline = article['MedlineCitation']
                article_data = medline['Article']

                # Estrazione MeSH Terms
                mesh_terms = []
                if 'MeshHeadingList' in medline:
                    mesh_terms = [m['DescriptorName'] for m in medline['MeshHeadingList']]

                # Estrazione Abstract
                abstract = ""
                if 'Abstract' in article_data:
                    abstract = " ".join(article_data['Abstract']['AbstractText'])

                parsed_results.append({
                    "PMID": str(medline['PMID']),
                    "Titolo": article_data['ArticleTitle'],
                    "Anno": article_data['Journal']['JournalIssue']['PubDate'].get('Year', 'N/A'),
                    "Journal": article_data['Journal']['Title'],
                    "MeSH Terms": ", ".join(mesh_terms[:5]) + ("..." if len(mesh_terms) > 5 else ""),
                    "Abstract": abstract[:500] + "...",
                    "URL": f"https://pubmed.ncbi.nlm.nih.gov/{medline['PMID']}/"
                })

            return pd.DataFrame(parsed_results)

        except Exception as e:
            return f"Errore durante la ricerca: {e}"

# Istanza dell'agente
agent = MedlineAgent()

# Esempio di ricerca profonda sull'immunoterapia nel cancro polmonare negli ultimi 2 anni
df_deep = agent.deep_search("lung cancer immunotherapy", max_results=5, years_back=2)

display(df_deep)
````

````python
!pip install -q biopython
import pandas as pd
from Bio import Entrez

class BibliographyAgent:
    def __init__(self, email="your_email@example.com"):
        Entrez.email = email

    def get_bibliography(self, query, max_results=5):
        print(f"Searching Medline for: {query}...")
        try:
            # 1. Search for IDs
            handle = Entrez.esearch(db="pubmed", term=query, retmax=max_results)
            record = Entrez.read(handle)
            handle.close()
            id_list = record["IdList"]

            if not id_list:
                return "No results found."

            # 2. Fetch details
            handle = Entrez.efetch(db="pubmed", id=",".join(id_list), retmode="xml")
            records = Entrez.read(handle)
            handle.close()

            results = []
            for article in records['PubmedArticle']:
                details = article['MedlineCitation']['Article']
                pmid = article['MedlineCitation']['PMID']

                # Extract Year
                year = details['Journal']['JournalIssue']['PubDate'].get('Year', 'N/A')

                # Extract Authors (first 3)
                authors = "None"
                if 'AuthorList' in details:
                    author_names = [f"{a.get('LastName', '')} {a.get('Initials', '')}" for a in details['AuthorList'][:3]]
                    authors = ", ".join(author_names)

                results.append({
                    "Year": year,
                    "Authors": authors,
                    "Title": details['ArticleTitle'],
                    "Journal": details['Journal']['Title'],
                    "Medline Link": f"https://pubmed.ncbi.nlm.nih.gov/{pmid}/"
                })

            return pd.DataFrame(results)

        except Exception as e:
            return f"An error occurred: {e}"

# Instance and example run
bib_agent = BibliographyAgent()
df_bib = bib_agent.get_bibliography("psychosis recovery intervention guidelines 2024", max_results=5)
display(df_bib)
````

### Analisi Caso Clinico: Eritromelalgia Secondaria

**Sintomi**: Bruciore, arrossamento (eritema) e aumento della temperatura cutanea alle estremità.
**Trigger**: Calore, esercizio fisico.
**Patologia di base**: Trombocitemia Essenziale (un disturbo mieloproliferativo).

#### Fisiopatologia
Nell'eritromelalgia associata a disturbi mieloproliferativi, il meccanismo è legato all'attivazione e aggregazione piastrinica intravascolare che causa infiammazione microvascolare e trombosi arteriolare.

#### Trattamento di Prima Scelta
- **Aspirina (ASA) a basse dosi**: È altamente efficace e rappresenta il trattamento di scelta. Spesso produce un sollievo drammatico dai sintomi entro poche ore.

#### Diagnosi Differenziale (Opzioni errate):
- **Nifedipina**: Un calcio-antagonista che causa vasodilatazione; può paradossalmente peggiorare l'eritromelalgia.
- **Gabapentin**: Usato per il dolore neuropatico, ma non tratta la causa vascolare sottostante in questo caso.
- **Metoprololo**: Beta-bloccante, non ha ruolo nel controllo dell'attivazione piastrinica.
- **Idrossiclorochina**: Usata nel LES o AR, non indicata per eritromelalgia da trombocitemia.

### Fisiopatologia dell'Anemia Mielofitica (Mieloftisi)

L'anemia mielofitica non è una malattia primitiva, ma la conseguenza di una **sostituzione del midollo osseo** da parte di tessuti non emopoietici o cellule anomale. Ecco i passaggi chiave del meccanismo:

#### 1. Infiltrazione e Occupazione di Spazio
Il midollo osseo viene invaso da:
- **Metastasi di tumori solidi** (frequenti: prostata, mammella, polmone, neuroblastoma).
- **Malattie ematologiche proliferative** (mielofibrosi primaria, leucemie, linfomi).
- **Malattie granulomatose** (tubercolosi, sarcoidosi) o accumulo lipidico (Gaucher).

#### 2. Alterazione della Barriera Midollo-Sangue
Normalmente, il rilascio di cellule dal midollo al sangue è strettamente regolato dai **sinusoidi midollari**. L'invasione tumorale distrugge l'architettura dei vasi sinusoidali e la rete di stroma (fibrosi reattiva). Questo causa un "leakage" (perdita) di cellule immaturate nel circolo periferico:
- **Eritroblasti** (precursori dei globuli rossi).
- **Mielociti/Metamielociti** (precursori dei globuli bianchi).
Questo fenomeno è noto come **reazione leucoeritroblastica**.

#### 3. Stress Meccanico e Dacriociti
I globuli rossi, nel tentativo di uscire attraverso i sinusoidi distorti o fibrotici, subiscono una deformazione meccanica permanente. Questo porta alla formazione dei **dacriociti** (cellule a lacrima), segno distintivo della sofferenza midollare.

#### 4. Ematopoiesi Extramidollare
Poiché il midollo osseo non è più in grado di sostenere l'emopoiesi, il corpo tenta di compensare riattivando la produzione di sangue in organi fetali:
- **Milza (Splenomegalia)**.
- **Fegato (Epatomegalia)**.
Tuttavia, l'ematopoiesi in questi organi è inefficiente e contribuisce ulteriormente all'immissione in circolo di cellule immature e morfologicamente anomale.

#### 5. Pancitopenia Progressiva
Infine, la riduzione della massa totale di tessuto emopoietico funzionale porta alla carenza di tutte le linee cellulari:
- **Anemia** (riduzione O2).
- **Leucopenia** (rischio infezioni).
- **Trombocitopenia** (rischio emorragie).

#### **2. Fisiopatologia della Ginecomastia nel Cirrotico**

### Automazione del Ciclo di Ricerca

Per automatizzare il processo, creiamo uno script `orchestrate_research.py`. Questo script simula il comportamento dell'agente core:
1. Esegue il benchmark.
2. Verifica i vincoli.
3. Decide se mantenere (`keep`) o scartare (`discard`) le modifiche.
4. Ripete il processo.

````python
import subprocess
import json
import time
import re

def run_automated_loop(iterations=2):
    print(f"--- Avvio Loop di Ricerca Automatica ({iterations} iterazioni) ---")
    for i in range(1, iterations + 1):
        print(f"\n[Iterazione {i}] Esecuzione benchmark...")
        result = subprocess.run(['python3', 'simulate_treatment.py'], capture_output=True, text=True)
        try:
            match = re.search(r'\{.*?\}', result.stdout)
            if match:
                metrics = json.loads(match.group())
                print(f"Metriche rilevate: {metrics}")
                if result.returncode == 0:
                    print(f"✅ Successo: Efficacia {metrics.get('efficacy_score')}")
                else:
                    print("❌ Fallimento: Vincoli violati.")
            else:
                print("⚠️ Errore: Nessun JSON trovato.")
                print(f"Stderr: {result.stderr}")
        except Exception as e:
            print(f"Errore: {e}")
        time.sleep(0.5)
    print("\n--- Ciclo di Ricerca Completato ---")

run_automated_loop(2)
````

## Sviluppo Agente Ricerca Profonda

### Subtask:
Creazione di una classe Python 'MedicalResearchAgent' capace di eseguire ricerche avanzate su PubMed filtrando per data, termini MeSH e pertinenza clinica.


**Reasoning**:
I will define the `MedicalResearchAgent` class with the `deep_search` method as specified in the instructions to enable advanced PubMed searching and parsing.



````python
import pandas as pd
from Bio import Entrez
from datetime import datetime

class MedicalResearchAgent:
    """
    An agent capable of performing deep searches on PubMed using NCBI Entrez APIs.
    """
    def __init__(self, email="agent@colab.research.local"):
        Entrez.email = email

    def deep_search(self, query, max_results=10, years_back=None, sort='relevance'):
        """
        Executes an advanced search on PubMed with filtering and metadata extraction.
        """
        print(f"Searching PubMed for: '{query}'...")

        # 1. Date Filtering
        if years_back:
            current_year = datetime.now().year
            date_query = f" AND ({current_year - years_back}:{current_year}[pdat])"
            query += date_query

        try:
            # 2. Search for IDs
            handle = Entrez.esearch(db="pubmed", term=query, retmax=max_results, sort=sort)
            record = Entrez.read(handle)
            handle.close()

            id_list = record.get("IdList", [])
            if not id_list:
                return pd.DataFrame()  # Return empty DF if no results

            # 3. Fetch Full Metadata (XML)
            handle = Entrez.efetch(db="pubmed", id=",".join(id_list), retmode="xml")
            records = Entrez.read(handle)
            handle.close()

            # 4. Parsing the XML structure
            findings = []
            for article in records.get('PubmedArticle', []):
                citation = article.get('MedlineCitation', {})
                article_info = citation.get('Article', {})

                # Basic Metadata
                title = article_info.get('ArticleTitle', 'N/A')
                journal = article_info.get('Journal', {}).get('Title', 'N/A')
                year = article_info.get('Journal', {}).get('JournalIssue', {}).get('PubDate', {}).get('Year', 'N/A')
                pmid = citation.get('PMID', 'N/A')

                # MeSH Terms extraction
                mesh_list = citation.get('MeshHeadingList', [])
                mesh_terms = [m['DescriptorName'] for m in mesh_list] if mesh_list else []

                # Abstract extraction
                abstract_data = article_info.get('Abstract', {}).get('AbstractText', [])
                abstract = " ".join(abstract_data) if isinstance(abstract_data, list) else str(abstract_data)

                findings.append({
                    "PMID": str(pmid),
                    "Title": title,
                    "Year": year,
                    "Journal": journal,
                    "MeSH Terms": ", ".join(mesh_terms),
                    "Abstract": abstract,
                    "URL": f"https://pubmed.ncbi.nlm.nih.gov/{pmid}/"
                })

            return pd.DataFrame(findings)

        except Exception as e:
            print(f"API Connection or Parsing Error: {e}")
            return pd.DataFrame()

# Instantiate the agent
research_agent = MedicalResearchAgent()
print("MedicalResearchAgent defined and ready.")
````

- **Moderata**: Diarrea che interferisce con le attività (es. costringe a modificare l'itinerario) ma senza segni di allarme.
- **Grave/Dissenteria**: Presenza di sangue nelle feci, febbre alta o grave disidratazione.

*   **Reidratazione**: La pietra miliare è il ripristino dei liquidi e degli elettroliti (Oral Rehydration Salts - ORS).

### 3. Razionale per l'Esclusione di Altre Terapie
*   **Antibiotici**: L'uso di antibiotici (es. Azitromicina o Fluorochinoloni) è raccomandato solo per la **diarrea grave**. Nei casi lievi/moderati, il rischio di effetti collaterali e lo sviluppo di resistenze antimicrobiche superano i benefici di una guarigione leggermente più rapida.
*   **Ospedalizzazione**: Non è indicata a meno che non vi sia una disidratazione incoercibile per via orale o sospetto di complicanze sistemiche.

### Conclusione
Per un paziente con sintomi lievi o moderati, senza febbre o sangue nelle feci, la condotta corretta è la **terapia di supporto con loperamide e liquidi**, evitando l'over-treatment antibiotico in linea con i principi di stewardship antimicrobica.

````python
from fpdf import FPDF

class TXAReport(FPDF):
    def header(self):
        self.set_font('Arial', 'B', 14)
        self.cell(0, 10, 'TXA Protocol Scalability Report: Urology Unit', 0, 1, 'C')
        self.ln(5)

    def chapter_title(self, title):
        self.set_font('Arial', 'B', 12)
        self.set_fill_color(200, 220, 255)
        self.cell(0, 6, title, 0, 1, 'L', 1)
        self.ln(4)

    def chapter_body(self, body):
        self.set_font('Arial', '', 11)
        self.multi_cell(0, 10, body)
        self.ln()

pdf = TXAReport()
pdf.add_page()

pdf.chapter_title('1. Clinical Efficacy Summary')
pdf.chapter_body('Statistical analysis of the 50-patient audit cohort confirms the efficacy of the new Tranexamic Acid (TXA) protocol. The mean hemostasis time was reduced to 447.35s (SD: 43.8s) compared to the historical baseline of 549.22s. This improvement of 101.87s is statistically significant (p < 0.001).')

pdf.chapter_title('2. Operational Compliance Gap')
pdf.chapter_body('Critical Issue: Despite drug efficacy, overall protocol compliance was recorded at only 40.0%. The main bottleneck identified is the failure to adhere to the 20-minute pre-incision administration window. Only 48% of cases achieved individual hemostasis targets below 450s due to these timing deviations.')

pdf.chapter_title('3. Scalability Recommendations')
pdf.chapter_body('- Mandatory "Time-Out" Integration: Link TXA administration to a hard stop in the pre-surgical checklist.\n- Staff Training: Launch targeted educational sessions focusing on the pharmacokinetic rationale for the 20-minute window.\n- Monitoring: Implement real-time digital tracking of administration-to-incision intervals for the next 200 cases.')

report_file = 'TXA_Scalability_Report.pdf'
pdf.output(report_file)
print(f'Report successfully generated: {report_file}')
````
