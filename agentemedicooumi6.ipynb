### 5. Aggiorna Node.js usando `nvm`

Installeremo `nvm` (Node Version Manager) e poi useremo `nvm` per installare e impostare una versione più recente di Node.js. La versione 18.x è una Long Term Support (LTS) e dovrebbe essere sufficiente per le tue dipendenze.

```python
# Installa nvm
!curl -o- https://raw.githubusercontent.com/nvm-sh/nvm/v0.39.1/install.sh | bash

# Carica nvm e installa Node.js LTS (versione 18.x) nella stessa sessione di shell
# È fondamentale che questi comandi siano eseguiti insieme per garantire che nvm sia caricato correttamente.
!bash -c 'export NVM_DIR="$HOME/.nvm" && [ -s "$NVM_DIR/nvm.sh" ] && . "$NVM_DIR/nvm.sh" && nvm install 18 && nvm use 18 && nvm alias default 18'

print("Node.js è stato aggiornato con nvm. Si prega di RIAVVIARE il runtime di Colab (Runtime -> Riavvia runtime) per applicare le modifiche e poi eseguire nuovamente le celle di verifica e avvio server.")
```

Dopo aver eseguito la cella qui sopra, **è fondamentale riavviare il runtime di Colab** (Runtime -> Riavvia runtime). Questo garantisce che `nvm` e la nuova versione di Node.js siano caricati correttamente nell'ambiente.

Una volta riavviato il runtime, esegui la prossima cella per verificare la versione di Node.js.

```python
# Verifica la versione di Node.js dopo il riavvio del runtime
!node -v
!npm -v
```
