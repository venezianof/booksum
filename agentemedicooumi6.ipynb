````python
import time
import requests

# Start ngrok tunnel in the background
get_ipython().system_raw('ngrok http 5173 &')

# Wait for ngrok to start and its local API to become available
# Implement a retry mechanism to robustly get the public URL
public_url = None
max_retries = 10
initial_delay = 5 # seconds

print("Attending the avvio di ngrok e recupero dell'URL pubblico...")

for i in range(max_retries):
    try:
        time.sleep(initial_delay)
        response = requests.get('http://localhost:4040/api/tunnels')
        tunnel_info = response.json()
        if 'tunnels' in tunnel_info and len(tunnel_info['tunnels']) > 0:
            public_url = tunnel_info['tunnels'][0]['public_url']
            break
    except requests.exceptions.ConnectionError as e:
        print(f"Tentativo {i+1}/{max_retries}: Connessione a ngrok API rifiutata. Riprovo...")
        if i == max_retries - 1:
            print(f"Impossibile connettersi a ngrok API dopo {max_retries} tentativi: {e}")
            raise

if public_url:
    print(f"Il tuo frontend Vite è disponibile al seguente URL pubblico: {public_url}")
else:
    print("Impossibile recuperare l'URL pubblico di ngrok. Assicurati che ngrok sia in esecuzione e configurato correttamente.")
````

### Classificazione Clinica dell'Ipertensione Polmonare (OMS)

L'ipertensione polmonare (IP) è classificata in 5 gruppi principali basati sulle cause fisiopatologiche e sulle modalità di trattamento:

1.  **Gruppo 1: Ipertensione Arteriosa Polmonare (IAP)**
    *   Include la forma idiopatica (causa ignota), quella ereditaria o associata ad altre condizioni (connettivopatie come la sclerodermia, infezione da HIV, ipertensione portale o cardiopatie congenite).
    *   Il problema risiede direttamente nelle piccole arterie polmonari che si restringono e si ispessiscono.

2.  **Gruppo 2: IP causata da malattie del cuore sinistro**
    *   È la forma più comune. È causata da problemi alle valvole cardiache (mitrale/aortica) o da insufficienza ventricolare sinistra.
    *   Il sangue "ristagna" nei polmoni perché il cuore sinistro non riesce a pomparlo efficacemente nel resto del corpo.

3.  **Gruppo 3: IP causata da malattie polmonari e/o ipossia**
    *   Include pazienti con BPCO (Broncopneumopatia Cronica Ostruttiva), fibrosi polmonare o disturbi del sonno (apnee notturne).
    *   La mancanza di ossigeno o il danno al tessuto polmonare causano la costrizione dei vasi sanguigni polmonari.

4.  **Gruppo 4: IP da ostruzioni delle arterie polmonari**
    *   Principalmente rappresentata dall'ipertensione polmonare cronica tromboembolica (CTEPH).
    *   È causata da vecchi coaguli di sangue (emboli) che non si sono sciolti e ostruiscono il flusso nelle arterie polmonari.

5.  **Gruppo 5: IP con meccanismi incerti o multifattoriali**
    *   Include condizioni sistemiche complesse come la sarcoidosi, l'anemia falciforme o disturbi metabolici che possono influenzare la pressione polmonare in modi diversi.

````python
import time
import requests

# Start ngrok tunnel in the background
get_ipython().system_raw('ngrok http 5173 &')

# Wait for ngrok to start and its local API to become available
# Implement a retry mechanism to robustly get the public URL
public_url = None
max_retries = 10
initial_delay = 5 # seconds

print("Attendo l'avvio di ngrok e recupero dell'URL pubblico...")

for i in range(max_retries):
    try:
        time.sleep(initial_delay)
        response = requests.get('http://localhost:4040/api/tunnels')
        tunnel_info = response.json()
        if 'tunnels' in tunnel_info and len(tunnel_info['tunnels']) > 0:
            public_url = tunnel_info['tunnels'][0]['public_url']
            break
    except requests.exceptions.ConnectionError as e:
        print(f"Tentativo {i+1}/{max_retries}: Connessione a ngrok API rifiutata. Riprovo...")
        if i == max_retries - 1:
            print(f"Impossibile connettersi a ngrok API dopo {max_retries} tentativi: {e}")
            raise

if public_url:
    print(f"Il tuo frontend Vite è disponibile al seguente URL pubblico: {public_url}")
else:
    print("Impossibile recuperare l'URL pubblico di ngrok. Assicurati che ngrok sia in esecuzione e configurato correttamente.")
````

````python
import requests
from bs4 import BeautifulSoup
import re

def search_medical_info(query):
    """
    Cerca informazioni su Wikipedia (versione italiana) con gestione degli errori.
    """
    search_url = "https://it.wikipedia.org/w/api.php"
    params = {
        "action": "query",
        "list": "search",
        "srsearch": query,
        "format": "json",
        "origin": "*"  # Necessario per evitare problemi di CORS o blocchi semplici
    }

    try:
        response = requests.get(search_url, params=params, timeout=10)
        response.raise_for_status() # Controlla se ci sono errori HTTP

        data = response.json()
        search_results = data.get("query", {}).get("search", [])

        if not search_results:
            return None

        return search_results[0]['title']
    except Exception as e:
        print(f"Errore durante la ricerca: {e}")
        return None

def get_page_content(title):
    """
    Estrae il testo principale da una pagina Wikipedia.
    """
    if not title:
        return "Titolo non valido."

    params = {
        "action": "query",
        "prop": "extracts",
        "exintro": True,
        "titles": title,
        "explaintext": True,
        "format": "json",
        "origin": "*"
    }
    url = "https://it.wikipedia.org/w/api.php"

    try:
        response = requests.get(url, params=params, timeout=10)
        response.raise_for_status()
        data = response.json()
        pages = data.get("query", {}).get("pages", {})

        for page_id in pages:
            return pages[page_id].get("extract", "Contenuto non disponibile.")
    except Exception as e:
        return f"Errore nel recupero del contenuto: {e}"

# Esecuzione della ricerca
query_utente = "estensione gomito gradi fisiologia"
titolo_pagina = search_medical_info(query_utente)

if titolo_pagina:
    contenuto = get_page_content(titolo_pagina)
    print(f"Ricerca per: {query_utente}")
    print(f"Fonte: Wikipedia - {titolo_pagina}\n")
    print(f"Estratto: {contenuto[:600]}...")
else:
    print("Impossibile trovare informazioni automatiche. Utilizzo conoscenze mediche predefinite.")
    print("L'estensione completa del gomito è solitamente definita come 0°. Un range di normalità comune include un'iperestensione fisiologica fino a 10°.")
````

2. Create a loop that will attempt to connect to the ngrok local API at `http://localhost:4040/api/tunnels` for a set number of retries (e.g., 10 to 15 attempts).

## Recupero URL Pubblico

### Subtask:
Attempt to retrieve the ngrok public URL from the local API to verify if the tunnel is accessible despite the process check findings.


**Reasoning**:
I will implement a Python script to poll the ngrok local API for the public URL, providing a final verification step as per the instructions.



````python
import requests
import time

print("--- Verifica Finale: Recupero URL tramite API ngrok ---")

api_url = "http://localhost:4040/api/tunnels"
public_url = None
max_attempts = 5

for i in range(max_attempts):
    try:
        response = requests.get(api_url, timeout=5)
        data = response.json()
        tunnels = data.get('tunnels', [])
        if tunnels:
            public_url = tunnels[0].get('public_url')
            print(f"✅ Tunnel trovato al tentativo {i+1}: {public_url}")
            break
        else:
            print(f"Tentativo {i+1}/{max_attempts}: Nessun tunnel attivo nell'API.")
    except requests.exceptions.ConnectionError:
        print(f"Tentativo {i+1}/{max_attempts}: Connessione all'API ngrok rifiutata (Porta 4040 non attiva).")
    except Exception as e:
        print(f"Tentativo {i+1}/{max_attempts}: Errore imprevisto: {e}")

    if i < max_attempts - 1:
        time.sleep(2)

if not public_url:
    print("\n❌ Conclusione: L'URL pubblico non è recuperabile. Il tunnel ngrok non è attivo.")
else:
    print(f"\n🚀 L'URL pubblico è: {public_url}")
````
