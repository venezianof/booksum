Model page: https://huggingface.co/lm-provers/QED-Nano

````python
```python
# Use a pipeline as a high-level helper
from transformers import pipeline

# Specifichiamo max_new_tokens per evitare il conflitto con max_length
pipe = pipeline("text-generation", model="lm-provers/QED-Nano")
messages = [
    {"role": "user", "content": "What is the Pythagorean theorem?"},
]
# Aggiungiamo il parametro qui per una risposta completa e senza warning
pipe(messages, max_new_tokens=256)
```
````

````python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("lm-provers/QED-Nano")
model = AutoModelForCausalLM.from_pretrained("lm-provers/QED-Nano")
messages = [
    {"role": "user", "content": "Give a practical example or analogy explaining how GFR decreases during a downstream obstruction, using the concept of hydrostatic pressure."},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)

# Using 256 tokens for a clear practical example
outputs = model.generate(**inputs, max_new_tokens=256)
print(tokenizer.decode(outputs[0][inputs["input_ids"].shape[-1]:]))
````

````python
from transformers import pipeline

# Inizializziamo il pipe se non è già presente
pipe = pipeline("text-generation", model="lm-provers/QED-Nano")

messages = [
    {"role": "user", "content": "Give a practical example or analogy explaining how GFR decreases during a downstream obstruction, using the concept of hydrostatic pressure."},
]

# Generazione con parametri diversi: temperature, top_p e repetition_penalty
outputs = pipe(
    messages,
    max_new_tokens=256,
    temperature=0.7,
    top_p=0.9,
    repetition_penalty=1.2,
    do_sample=True
)

print(outputs[0]['generated_text'][-1]['content'])
````
