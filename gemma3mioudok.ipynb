### 🏥 Deep Medical Research & Therapy Agent
Questo agente è progettato per:
1.  **Analisi Clinica**: Estrarre informazioni chiave da sintomi o referti.
2.  **Ricerca Scientifica**: Sintetizzare protocolli basati su letteratura medica.
3.  **Pianificazione Terapeutica**: Suggerire dosaggi, monitoraggi e controindicazioni.

```python
def medical_agent_query(clinical_case, focus_area='therapy'):
    """
    Funzione core per l'agente di ricerca medica.
    """
    system_prompt = """Sei un assistente IA specializzato in medicina avanzata.
    Il tuo compito è fornire analisi basate sull'evidenza scientifica.
    Usa un linguaggio tecnico preciso, cita protocolli standard (come linee guida ESC, AHA, AIFA)
    e specifica sempre le avvertenze sui rischi."""

    prompt = f"""ANALISI CASO CLINICO:
    {clinical_case}

    RICHIESTA: Genera una ricerca profonda su {focus_area} e una proposta di terapia dettagliata.
    Includi: Meccanismo d'azione, posologia standard, effetti collaterali e interazioni farmacologiche."""

    response = llm.create_chat_completion(
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": prompt}
        ],
        temperature=0.2, # Bassa temperatura per maggiore precisione medica
        max_tokens=2048
    )
    return response['choices'][0]['message']['content']
```

```python
clinical_case_user = """Un uomo di 72 anni si presenta al pronto soccorso con dispnea acuta, tachicardia e lieve dolore toracico insorto 2 ore prima. Non ha storia di neoplasia, recenti interventi chirurgici o immobilizzazione prolungata. Il punteggio di Wells è basso. Viene eseguito un dosaggio del D-dimero con metodo ELISA: il risultato è 680 ng/mL."""

query_result = medical_agent_query(
    clinical_case=clinical_case_user,
    focus_area="interpretazione D-dimero ed embolia polmonare nel paziente anziano e le conseguenti decisioni cliniche in relazione al cutoff aggiustato per età"
)

print(query_result)
```

### 📄 Aggiornamento Linee Guida ESC 2024/2025: Fibrillazione Atriale
Le nuove linee guida sostituiscono il precedente schema 'CC/ABC' con il framework **CARE**:
1. **C**omorbidity management (Gestione delle comorbidità)
2. **A**voidance of stroke and thromboembolism (Prevenzione dell'ictus)
3. **R**ate and rhythm control (Controllo della frequenza e del ritmo)
4. **E**valuation and reassessment (Valutazione e rivalutazione continua)

```python
!pip install llama-cpp-python

from llama_cpp import Llama

# Inizializzazione del modello
llm = Llama.from_pretrained(
    repo_id="Andycurrent/Gemma-3-4B-VL-it-Gemini-Pro-Heretic-Uncensored-Thinking_GGUF",
    filename="Gemma-3-4B-VL-it-Gemini-Pro-Heretic-Uncensored-Thinking_F16.gguf",
    n_ctx=4096,
    n_gpu_layers=-1
)

def medical_agent_query(clinical_case, focus_area='therapy'):
    system_prompt = "Sei un assistente IA specializzato in medicina avanzata. Usa un linguaggio tecnico preciso e cita protocolli ESC 2024."
    prompt = f"ANALISI CASO CLINICO:\n{clinical_case}\n\nRICHIESTA: {focus_area}"
    response = llm.create_chat_completion(
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": prompt}
        ],
        temperature=0.1
    )
    return response['choices'][0]['message']['content']

# Esecuzione della richiesta specifica
focus_esc = "Linee guida ESC 2024/2025 Fibrillazione Atriale, framework CARE, nuovi score di rischio clinico, indicazioni all'ablazione e gestione anticoagulante."
risultato_esc = medical_agent_query(
    clinical_case="Paziente con nuova diagnosi di fibrillazione atriale parossistica, iperteso, con necessità di inquadramento secondo i protocolli più recenti.",
    focus_area=focus_esc
)

print(risultato_esc)
```

### 🔄 Confronto: ESC 2020 (CC/ABC) vs ESC 2024 (CARE)

| Caratteristica | Modello ESC 2020 (ABC) | Modello ESC 2024 (CARE) |
| :--- | :--- | :--- |
| **Focus Principale** | Gestione sequenziale dei sintomi e del rischio | Gestione olistica e dinamica (Rivalutazione) |
| **Framework** | **CC/ABC** | **CARE** |
| **Prevenzione Ictus** | Score CHA₂DS₂-VASc standard | Score di rischio clinico rifiniti e prevenzione ('A') |
| **Controllo Ritmo** | Ablazione come seconda linea (spesso) | Indicazione precoce all'ablazione ('R') |
| **Follow-up** | Basato su necessità clinica | Valutazione e rivalutazione sistematica ('E') |

```python
confronto_prompt = """
Effettua un confronto approfondito tra le linee guida ESC 2020 e ESC 2024 per la Fibrillazione Atriale.
In particolare, analizza come il framework CARE (2024) migliori o modifichi l'approccio ABC (2020).
Focus su:
1. Gestione delle comorbidità (C)
2. Evoluzione degli score di rischio (A)
3. Timing dell'ablazione transcatetere (R)
4. L'importanza del pilastro 'Evaluation and reassessment' (E).
"""

risultato_confronto = medical_agent_query(
    clinical_case="Evoluzione storica dei protocolli ESC per AF.",
    focus_area=confronto_prompt
)

print(risultato_confronto)
```

#### Esempio di utilizzo: Ricerca su Diabete di Tipo 2 e Nuove Terapie

```python
caso_clinico = "Paziente 55 anni, BMI 32, HbA1c 8.2% nonostante Metformina 1000mg x2. Anamnesi di ipertensione controlled."
risultato_ricerca = medical_agent_query(caso_clinico, focus_area="SGLT2 inibitori e GLP-1 agonisti")

print(risultato_ricerca)
```

### 👁️ Analisi Multimodale (Visione)
Il modello Gemma-3 4B VL può elaborare immagini. Di seguito un esempio di come inviare un'immagine tramite URL per l'analisi.

```python
# Esempio di analisi di un'immagine (es. un grafico clinico o una foto di un farmaco)
image_url = "https://cdn.britannica.com/61/93061-050-99147DCE/Statue-of-Liberty-Island-New-York-Bay.jpg"

response = llm.create_chat_completion(
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "Descrivi cosa vedi in questa immagine con un linguaggio professionale."},
                {
                    "type": "image_url",
                    "image_url": {"url": image_url}
                }
            ]
        }
    ]
)

print(response['choices'][0]['message']['content'])
```
