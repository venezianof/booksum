### Interruzione del Trattamento con Eparina o Fondaparinux in Embolia Polmonare

La risposta corretta è: **Dopo cinque giorni di trattamento, se due valori consecutivi di INR misurati a 24 ore di distanza sono >2.**

#### Razionale Clinico:
Il trattamento dell'embolia polmonare (EP) inizia con anticoagulanti parenterali a rapida azione (come l'eparina o il fondaparinux). Contemporaneamente o poco dopo, si avvia la terapia con un antagonista della vitamina K (VKA, come il warfarin) per la terapia a lungo termine. È fondamentale che vi sia un periodo di **overlap** tra i due farmaci per garantire un'anticoagulazione efficace e stabile.

Le linee guida cliniche raccomandano di:
*   Mantenere l'anticoagulazione parenterale per un **minimo di 5 giorni**.
*   Discontinuare l'eparina o il fondaparinux solo dopo che l'INR (International Normalized Ratio), misurato mentre il paziente assume il VKA, abbia raggiunto un valore terapeutico (generalmente tra 2.0 e 3.0) per **almeno due valori consecutivi misurati a distanza di 24 ore**.

Questo approccio assicura che il VKA abbia avuto il tempo di esplicare pienamente il suo effetto anticoagulante, prevenendo così il rischio di recidive tromboemboliche una volta interrotto il farmaco parenterale.

#### Perché le altre opzioni sono errate:
*   **Non è mai possibile interrompere il trattamento con eparina fondaparinux fino alla completa risoluzione dell'EP:** Falso. I farmaci parenterali vengono interrotti una volta che l'anticoagulazione orale è efficace.
*   **Immediatamente dopo l'inizio del trattamento con antagonisti della vitamina K, senza necessità di valutare l'INR:** Falso. È richiesto un periodo di overlap e la conferma del raggiungimento dell'INR terapeutico.
*   **Dopo un mese di trattamento, indipendentemente dai valori di INR:** Falso. La durata è un fattore, ma l'INR terapeutico è il criterio chiave per la transizione.
*   **Dopo tre giorni di trattamento, se il paziente non presenta più sintomi di EP:** Falso. La risoluzione dei sintomi non è sufficiente; il criterio principale è l'INR terapeutico, e la durata minima è generalmente di 5 giorni.

### Fisiopatologia Cardiaca: Cardiomiopatia Restrittiva

La risposta corretta è: **Cardiomiopatia restrittiva**.

#### Razionale Clinico:
La **cardiomiopatia restrittiva** è caratterizzata da una spiccata rigidità delle pareti ventricolari (diminuita compliance). Questo comporta che:
*   Anche piccoli aumenti del volume ematico durante il riempimento (diastole) causano un **aumento sproporzionato della pressione** all'interno della cavità.
*   La funzione sistolica (contrattilità) è spesso preservata nelle fasi iniziali, ma il problema risiede nell'incapacità del ventricolo di distendersi per accogliere il sangue.
*   Le cause possono essere idiopatiche o secondarie a malattie infiltrative (come l'amiloidosi o la sarcoidosi) o fibrosi endomiocardica.

#### Perché le altre opzioni sono errate:
*   **Ipertrofia ventricolare sinistra:** È un aumento dello spessore delle pareti (spesso dovuto a ipertensione), ma non definisce di per sé il fenomeno della restrizione diastolica pura descritto nel quesito.
*   **Pericardite adesiva:** Sebbene causi un ostacolo al riempimento (costrizione), il problema è legato al pericardio ispessito che 'schiaccia' il cuore dall'esterno, non alla rigidità intrinseca della parete ventricolare.
*   **Sindrome di Brugada:** È una canalopatia aritmogena ereditaria che predispone alla morte improvvisa, ma non comporta alterazioni strutturali della rigidità di parete.
*   **Cardiomiopatia ipertrofica:** È caratterizzata da un ispessito miocardio (spesso asimmetrico) e disfunzione diastolica, ma la definizione di 'sproporzionato aumento pressorio per minimo riempimento dovuto a rigidità' è il segno cardine e patognomonico della forma *restrittiva*.

### Riassunto dei Pattern Radiologici HRTC (Interstiziopatie)

| Pattern | Distribuzione Prevalente | Caratteristiche Chiave | Associazione Tipica |
| :--- | :--- | :--- | :--- |
| **UIP** (Usual Interstitial Pneumonia) | Basale e Subpleurica | **Honeycombing** (nido d'ape), reticolazione, bronchiettasie da trazione. | Fibrosi Polmonare Idiopatica (IPF), Asbestosi. |

1.  **Honeycombing:** È il segno cardine della **UIP**. Se assente, la diagnosi si sposta verso NSIP o UIP indeterminata.
2.  **Vetro Smerigliato (Ground-glass):** Se domina il quadro senza fibrosi distruttiva, si sospetta **NSIP** o **DIP**.
3.  **Consolidamenti:** Se sono presenti aree di 'riempimento' alveolare piuttosto che reticolazione, il sospetto principale è la **COP**.

### Approfondimento sul Circuito di Papez

Il **circuito di Papez** è un sistema neuronale complesso situato nel lobo limbico, descritto originariamente nel 1937 dal neuroanatomista James Papez. Inizialmente proposto come la base anatomica delle emozioni, oggi sappiamo che gioca un ruolo cruciale anche nel consolidamento della memoria a lungo termine.

#### 1. Il Percorso Anatomico Passo-Passo
Il circuito è organizzato in una struttura chiusa (loop) che collega diverse aree del proencefalo:

*   **Ippocampo**: Il punto di partenza. Elabora le informazioni e le invia tramite il **fornice**.
*   **Corpi Mamillari**: Ricevono i segnali dal fornice. Si trovano nell'ipotalamo posteriore.
*   **Fascio Mamillo-Talamico**: Un fascio di fibre che proietta dai corpi mamillari ai nuclei anteriori del talamo.
*   **Nuclei Anteriori del Talamo**: Fungono da stazione di rilievo, proiettando verso la corteccia cingolata.
*   **Corteccia del Cingolo**: Qui avviene l'esperienza conscia delle emozioni. L'informazione viaggia lungo il cingolo verso il giro paraippocampale.
*   **Giro Paraippocampale (Area Entorinale)**: Chiude il cerchio proiettando nuovamente all'ippocampo.

#### 2. Funzioni Principali
Sebbene Papez credesse che il circuito servisse esclusivamente a generare l'esperienza emotiva (sentimento), la ricerca moderna ha ampliato questa visione:

*   **Regolazione Emozionale**: Contribuisce all'integrazione tra stimoli sensoriali e risposte affettive.
*   **Memoria Episodica**: È fondamentale per trasformare i ricordi a breve termine in ricordi a lungo termine. Danni in qualsiasi punto di questo circuito (es. sindrome di Korsakoff nei corpi mamillari) possono causare amnesia anterograda.

#### 3. Evoluzione del concetto: Il Sistema Limbico
Il circuito di Papez è stato successivamente integrato nel concetto più ampio di **Sistema Limbico** proposto da Paul MacLean, che ha incluso altre strutture vitali come l'**amigdala** (centrale per la paura) e la **corteccia prefrontale** (per il controllo esecutivo delle emozioni).

### Risposta al quesito di Anatomia: Articolazione del Gomito

La risposta corretta è: **Superfici articolari ricoperte da cartilagine ialina**.

#### Razionale Anatomico:
L'articolazione del gomito è un'articolazione sinoviale complessa che comprende tre giunzioni all'interno di un'unica capsula articolare:
1.  **Humeroulnare** (ginglimo angolare).
2.  **Humeroradiale** (condiloidea).
3.  **Radioulnare prossimale** (ginglimo laterale).

In tutte queste giunzioni, le superfici ossee a contatto (troclea e condilo dell'omero, incisura trocleare dell'ulna e capitello del radio) sono rivestite da **cartilagine ialina** (o cartilagine articolare).

#### Funzioni della Cartilagine Ialina:
*   **Riduzione dell'attrito**: Permette lo scivolamento fluido delle superfici ossee durante la flesso-estensione e la prono-supinazione.
*   **Distribuzione del carico**: Ammortizza le forze di compressione che si esercitano sull'articolazione.

#### Perché le altre opzioni sono errate:
*   **Prive di cartilagine**: Impossibile per un'articolazione mobile (diartrosi); porterebbe a una rapida erosione ossea e dolore estremo.
*   **Fibrocartilagine**: La fibrocartilagine si trova in articolazioni sottoposte a stress da trazione o pressione molto elevata (es. dischi intervertebrali, menischi, sinfisi pubica), ma non riveste le superfici del gomito.
*   **Cartilagine fibro-elastica**: Si trova in strutture come il padiglione auricolare o l'epiglottide, non nelle superfici articolari.
*   **Collegate direttamente ai legamenti**: Sebbene i legamenti stabilizzino l'articolazione, essi non sostituiscono il rivestimento cartilagineo delle superfici di contatto osseo.

## Risposta al quesito medico: Durata dell'Astinenza da Oppiacei a Breve Durata d'Azione

La durata tipica dei sintomi acuti di astinenza da oppiacei per prodotti a breve durata d'azione è: **3-7 giorni**.

### Razionale Clinico:
L'astinenza da oppiacei a breve durata d'azione, come l'eroina o alcuni oppiacei da prescrizione (es. ossicodone a rilascio immediato), segue un decorso temporale ben definito a causa della rapida eliminazione del farmaco dall'organismo.

*   **Esordio dei sintomi:** Generalmente entro 6-12 ore dall'ultima dose.
*   **Picco dei sintomi:** Tipicamente raggiunto entro 48-72 ore (2-3 giorni).

### Perché le altre opzioni sono errate:

*   **1-2 giorni:** Sebbene i sintomi possano essere intensi in questo periodo, la loro durata complessiva è solitamente maggiore.

*   **12-24 ore:** Questo è il periodo in cui i sintomi sono in pieno sviluppo o raggiungono il picco, non la durata totale dell'astinenza acuta.
*   **2-3 mesi:** Questa è una durata eccessivamente lunga per la fase acuta dell'astinenza da oppiacei a breve durata d'azione e potrebbe indicare sintomi cronici, complicazioni o ricaduta, non il decorso tipico dell'astinenza acuta.

#### Perché è la scelta corretta?
La paziente soffre di insonnia nella **seconda metà della notte**, il che indica un problema di **mantenimento del sonno**. Questo tipo di insonnia richiede un farmaco con un'**emivita sufficientemente lunga** da coprire le ore restanti di sonno, senza però causare eccessiva sedazione diurna (hangover).

Questa durata d'azione lo rende adatto per il mantenimento del sonno, in quanto può coprire la seconda metà della notte. Sebbene le benzodiazepine debbano essere usate con cautela negli anziani (a causa del rischio di cadute, confusione e tolleranza/dipendenza), tra le opzioni con una lunga emivita è spesso considerata più gestibile di altre, se prescritta a dosaggi minimi efficaci.

#### Perché le altre opzioni sono meno adatte?
*   **Zolpidem**: Ha un'**emivita molto breve (circa 2-3 ore)**. È indicato principalmente per l'insonnia da **difficoltà di addormentamento**, non per il mantenimento del sonno nella seconda metà della notte, poiché verrebbe metabolizzato troppo rapidamente.
*   **Doxilammina**: È un antistaminico con proprietà sedative e un'**emivita relativamente lunga (circa 10-12 ore o più)**. Nonostante l'emivita lunga, la sua tossicità in una paziente anziana lo rende meno adatto.
*   **Zopiclone**: Ha un'**emivita breve-intermedia (circa 5-6 ore)**. Potrebbe aiutare parzialmente il mantenimento, ma l'emivita è più breve di quella del lormetazepam e potrebbe non essere sufficiente per coprire l'intera seconda metà della notte in un sonno di 7-8 ore.
*   **Melatonina**: Ha un'**emivita molto breve (circa 20-60 minuti)**. Viene utilizzata principalmente per disturbi del ritmo circadiano o insonnia da addormentamento. Non è efficace per il mantenimento del sonno a causa della sua rapida eliminazione.

In sintesi, il Lormetazepam è l'opzione che meglio bilancia l'esigenza di una molecola con un'emivita "più lunga" per il mantenimento del sonno, con un profilo di rischio gestibile (sebbene richieda cautela) in una paziente anziana, rispetto alla doxilammina che presenta rischi maggiori.

# 🎯 Part 1: Supervised Fine-Tuning (SFT)

SFT teaches the model to follow instructions by training on input-output pairs (instruction vs response). This is the foundation for creating instruction-following models.

## Load an SFT Dataset

We will use [HuggingFaceTB/smoltalk](https://huggingface.co/datasets/HuggingFaceTB/smoltalk), limiting ourselves to the first 5k samples for brevity. Feel free to change the limit by changing the slicing index in the parameter `split`.

```python
from datasets import load_dataset

print("📥 Loading SFT dataset...")
train_dataset_sft = load_dataset("HuggingFaceTB/smoltalk", "all", split="train[:5000]")
eval_dataset_sft = load_dataset("HuggingFaceTB/smoltalk", "all", split="test[:500]")

print("✅ SFT Dataset loaded:")
print(f"   📚 Train samples: {len(train_dataset_sft)}")
print(f"   🧪 Eval samples: {len(eval_dataset_sft)}")
print(f"\n📝 Single Sample: {train_dataset_sft[0]['messages']}")
```

## Launch Training

We are now ready to launch an SFT run with `SFTTrainer`, feel free to modify `SFTConfig` to play around with different configurations.



```python
from trl import SFTConfig, SFTTrainer

sft_config = SFTConfig(
    output_dir="./lfm2-sft",
    num_train_epochs=1,
    per_device_train_batch_size=1,
    learning_rate=5e-5,
    lr_scheduler_type="linear",
    warmup_steps=100,
    warmup_ratio=0.2,
    logging_steps=10,
    save_strategy="epoch",
    eval_strategy="epoch",
    load_best_model_at_end=True,
    report_to=None,
    bf16=False # <- not all colab GPUs support bf16
)

print("🏗️  Creating SFT trainer...")
sft_trainer = SFTTrainer(
    model=model,
    args=sft_config,
    train_dataset=train_dataset_sft,
    eval_dataset=eval_dataset_sft,
    processing_class=tokenizer,
)

print("\n🚀 Starting SFT training...")
sft_trainer.train()

print("🎉 SFT training completed!")

sft_trainer.save_model()
print(f"💾 SFT model saved to: {sft_config.output_dir}")
```

# 🎛️ Part 2: LoRA + SFT (Parameter-Efficient Fine-tuning)

LoRA (Low-Rank Adaptation) allows efficient fine-tuning by only training a small number of additional parameters. Perfect for limited compute resources!


## Wrap the model with PEFT

We specify target modules that will be finetuned while the rest of the models weights remains frozen. Feel free to modify the `r` (rank) value:
- higher -> better approximation of full-finetuning
- lower -> needs even less compute resources

```python
from peft import LoraConfig, get_peft_model, TaskType

GLU_MODULES = ["w1", "w2", "w3"]
MHA_MODULES = ["q_proj", "k_proj", "v_proj", "out_proj"]
CONV_MODULES = ["in_proj", "out_proj"]

lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=8,  # <- lower values = fewer parameters
    lora_alpha=16,
    lora_dropout=0.1,
    target_modules=GLU_MODULES + MHA_MODULES + CONV_MODULES,
    bias="none",
    modules_to_save=None,
)

lora_model = get_peft_model(model, lora_config)
lora_model.print_trainable_parameters()

print("✅ LoRA configuration applied!")
print(f"🎛️  LoRA rank: {lora_config.r}")
print(f"📊 LoRA alpha: {lora_config.lora_alpha}")
print(f"🎯 Target modules: {lora_config.target_modules}")
```

## Launch Training

Now ready to launch the SFT training, but this time with the LoRA-wrapped model

```python
from trl import SFTConfig, SFTTrainer

lora_sft_config = SFTConfig(
    output_dir="./lfm2-sft-lora",
    num_train_epochs=1,
    per_device_train_batch_size=1,
    learning_rate=5e-5,
    lr_scheduler_type="linear",
    warmup_steps=100,
    warmup_ratio=0.2,
    logging_steps=10,
    save_strategy="epoch",
    eval_strategy="epoch",
    load_best_model_at_end=True,
    report_to=None,
)

print("🏗️  Creating LoRA SFT trainer...")
lora_sft_trainer = SFTTrainer(
    model=lora_model,
    args=lora_sft_config,
    train_dataset=train_dataset_sft,
    eval_dataset=eval_dataset_sft,
    processing_class=tokenizer,
)

print("\n🚀 Starting LoRA + SFT training...")
lora_sft_trainer.train()

print("🎉 LoRA + SFT training completed!")

lora_sft_trainer.save_model()
print(f"💾 LoRA model saved to: {lora_sft_config.output_dir}")
```

### Funzionalità chiave dell'agente di ricerca medica:

1.  **Ricerca di diagnosi:** L'agente deve essere in grado di ricercare e presentare informazioni dettagliate su diagnosi mediche, inclusi sintomi, eziologia, patogenesi e criteri diagnostici.
2.  **Identificazione di farmaci:** Deve poter identificare farmaci pertinenti per specifiche condizioni, fornendo dettagli su meccanismi d'azione, dosaggi, vie di somministrazione e formulazioni.
3.  **Analisi degli effetti collaterali:** L'agente dovrà ricercare e riassumere gli effetti collaterali noti dei farmaci, distinguendo tra comuni, rari e gravi, e le loro interazioni.
4.  **Ricerca di studi clinici:** Capacità di cercare e filtrare studi clinici rilevanti (in corso, completati, futuri) per condizioni specifiche, inclusi i risultati chiave e le metodologie.
5.  **Ricerca di linee guida cliniche:** Rintracciamento e sintesi delle linee guida cliniche più recenti e autorevoli per la gestione delle malattie.

### Tipi di informazioni da ricercare e elaborare:

*   **Malattie e condizioni:** Definizione, epidemiologia, fattori di rischio, diagnosi differenziale, stadiazione.
*   **Farmaci:** Nomi generici e commerciali, classi farmacologiche, indicazioni, controindicazioni, avvertenze, farmacocinetica e farmacodinamica.
*   **Trattamenti e terapie:** Opzioni terapeutiche (farmacologiche, chirurgiche, non farmacologiche), protocolli, efficacia e sicurezza comparativa.
*   **Evidenze scientifiche:** Risultati di metanalisi, revisioni sistematiche, studi randomizzati controllati (RCT).

### Fonti di dati mediche autorevoli e pertinenti:

*   **PubMed/MEDLINE:** Per la letteratura scientifica peer-reviewed.
*   **ClinicalTrials.gov / EU Clinical Trials Register:** Per i registri di studi clinici.
*   **Linee guida ufficiali:** NICE, EMA, FDA, WHO, società scientifiche nazionali e internazionali.
*   **Database farmacologici:** DrugBank, RxList, AIFA (per l'Italia).
*   **Database di effetti collaterali:** VigiBase, SIDER.

### Formati di output desiderati:

*   **Testo riassuntivo:** Sintesi concisa e facilmente comprensibile delle informazioni richieste.
*   **Elenchi puntati:** Per presentare rapidamente fatti chiave o opzioni.
*   **Tabelle comparative:** Per confrontare farmaci, trattamenti o diagnostici (es. efficacia, effetti collaterali, costi).
*   **Grafici:** Per visualizzare dati epidemiologici, trend di efficacia o distribuzione degli effetti collaterali (se possibile).
*   **Riferimenti bibliografici:** Link diretti alle fonti originali.

### Criteri per le raccomandazioni terapeutiche:

*   **Basate su evidenze scientifiche:** Le raccomandazioni devono essere supportate dai più alti livelli di evidenza disponibili.
*   **Aggiornate:** Utilizzare sempre le informazioni più recenti e le linee guida aggiornate.
*   **Personalizzate:** Considerare parametri specifici del paziente (età, comorbilità, allergie, farmaci concomitanti) per modulare le raccomandazioni.
*   **Trasparenza:** Indicare chiaramente il livello di evidenza e le fonti su cui si basano le raccomandazioni.
*   **Considerazione del rapporto rischio/beneficio:** Valutare attentamente i potenziali benefici e rischi delle terapie proposte.

Questo può includere modelli di Natural Language Processing (NLP) per la comprensione del testo medico, modelli di apprendimento automatico per l'analisi dei dati e modelli di ragionamento per la formulazione di raccomandazioni. To ensure the base model and its tokenizer are properly loaded and available in the current kernel session, I will execute the cell that performs this initialization.



```python
from transformers import AutoTokenizer, AutoModelForCausalLM
from IPython.display import display, HTML, Markdown
import torch

# Select a model to fine-tune from the list
lfm_models = [
    "LiquidAI/LFM2.5-1.2B-Instruct",
    "LiquidAI/LFM2.5-1.2B-JP",
    "LiquidAI/LFM2-8B-A1B",
    "LiquidAI/LFM2-2.6B-Exp",
    "LiquidAI/LFM2-2.6B",
    "LiquidAI/LFM2-700M",
    "LiquidAI/LFM2-350M",
]

# Model to fine-tune
model_id = "LiquidAI/LFM2.5-1.2B-Instruct"

print("📚 Loading tokenizer...")
tokenizer = AutoTokenizer.from_pretrained(model_id)

print("🧠 Loading model...")
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    device_map="auto",
    torch_dtype="bfloat16",
#   attn_implementation="flash_attention_2" <- uncomment on compatible GPU
)

print("✅ Local model loaded successfully!")
print(f"🔢 Parameters: {model.num_parameters():,}")
print(f"📖 Vocab size: {len(tokenizer)}")
print(f"💾 Model size: ~{model.num_parameters() * 2 / 1e9:.1f} GB (bfloat16)")
```

**Reasoning**:
The new subtask requires preparing a dataset for fine-tuning. The notebook contains a cell that loads the `HuggingFaceTB/smoltalk` dataset, which will be used for this purpose.



```python
from datasets import load_dataset

print("📥 Loading SFT dataset...")
train_dataset_sft = load_dataset("HuggingFaceTB/smoltalk", "all", split="train[:5000]")
eval_dataset_sft = load_dataset("HuggingFaceTB/smoltalk", "all", split="test[:500]")

print("✅ SFT Dataset loaded:")
print(f"   📚 Train samples: {len(train_dataset_sft)}")
print(f"   🧪 Eval samples: {len(eval_dataset_sft)}")
print(f"\n📝 Single Sample: {train_dataset_sft[0]['messages']}")
```

**Reasoning**:
The new subtask requires defining a fine-tuning strategy, specifically applying a PEFT technique like LoRA. The notebook contains a cell that sets up the `LoraConfig` and applies it to the previously loaded base model, creating `lora_model`.



```python
from peft import LoraConfig, get_peft_model, TaskType

GLU_MODULES = ["w1", "w2", "w3"]
MHA_MODULES = ["q_proj", "k_proj", "v_proj", "out_proj"]
CONV_MODULES = ["in_proj", "out_proj"]

lora_config = LoraConfig(
    task_type=TaskType.CAUSAL_LM,
    inference_mode=False,
    r=8,  # <- lower values = fewer parameters
    lora_alpha=16,
    lora_dropout=0.1,
    target_modules=GLU_MODULES + MHA_MODULES + CONV_MODULES,
    bias="none",
    modules_to_save=None,
)

lora_model = get_peft_model(model, lora_config)
lora_model.print_trainable_parameters()

print("✅ LoRA configuration applied!")
print(f"🎛️  LoRA rank: {lora_config.r}")
print(f"📊 LoRA alpha: {lora_config.lora_alpha}")
print(f"🎯 Target modules: {lora_config.target_modules}")
```

**Reasoning**:
The next step is to launch the fine-tuning process. The notebook contains a cell that sets up the `SFTConfig` for the LoRA + SFT training and then initiates the training using the `lora_model` and the prepared datasets. This directly addresses the instructions to define the fine-tuning strategy and launch the training.



```python
from trl import SFTConfig, SFTTrainer

lora_sft_config = SFTConfig(
    output_dir="./lfm2-sft-lora",
    num_train_epochs=1,
    per_device_train_batch_size=1,
    learning_rate=5e-5,
    lr_scheduler_type="linear",
    warmup_steps=100,
    warmup_ratio=0.2,
    logging_steps=10,
    save_strategy="epoch",
    eval_strategy="epoch",
    load_best_model_at_end=True,
    report_to=None,
    bf16=False # Explicitly setting bf16 to False to avoid unsupported bf16/gpu error
)

print("🏗️  Creating LoRA SFT trainer...")
lora_sft_trainer = SFTTrainer(
    model=lora_model,
    args=lora_sft_config,
    train_dataset=train_dataset_sft,
    eval_dataset=eval_dataset_sft,
    processing_class=tokenizer,
)

print("\n🚀 Starting LoRA + SFT training...")
lora_sft_trainer.train()

print("🎉 LoRA + SFT training completed!")

lora_sft_trainer.save_model()
print(f"💾 LoRA model saved to: {lora_sft_config.output_dir}")
```

### Primary Goal of Psychotherapy for Schizoid Personality Disorder

The primary goal of psychotherapy for Schizoid Personality Disorder (SPD) is to **help the individual connect with others in ways that feel safe and manageable.**

### Explanation for the Correct Choice

This is the correct choice because SPD is characterized by a pervasive pattern of detachment from social relationships and a restricted range of emotional expression. Individuals with SPD often have little desire for close relationships, prefer solitary activities, and are largely indifferent to praise or criticism. Their fear of intimacy and deep-seated discomfort with emotional closeness make traditional social interaction highly anxiety-provoking. Therefore, the therapeutic aim is not to force them into highly intimate or emotionally intense relationships, but rather to gradually develop social skills and coping mechanisms that allow them to engage in social connections at a level that feels comfortable and secure for them. The focus is on building a sense of safety within relationships and managing the anxiety that can arise from social interaction, rather than dramatically altering their fundamental personality traits.

### Why Other Options Are Incorrect

*   **Addressing delusions and hallucinations:** This is **incorrect** because delusions and hallucinations are characteristic symptoms of psychotic disorders like schizophrenia, not Schizoid Personality Disorder. SPD involves detachment and emotional flatness, but not a break from reality or a disturbance in thought content typically seen in psychosis. While individuals with SPD might occasionally experience brief psychotic episodes under extreme stress, this is not a primary, defining feature or therapeutic target for the disorder itself.

*   **Reducing symptoms of anxiety and depression:** While individuals with SPD may experience secondary anxiety or depression, especially when faced with social demands or isolation, reducing these symptoms is typically a *secondary* goal, not the *primary* one. The core of SPD lies in the pattern of detachment and restricted emotionality. Addressing these underlying personality traits and improving social functioning in a safe way is paramount; alleviation of anxiety and depression would often follow as a result of improved coping and a greater sense of security in their lives.

*   **Promoting greater emotional intensity and expression:** This is **incorrect** as a primary goal because forcing emotional intensity or expression often clashes with the fundamental nature of SPD. Individuals with SPD genuinely have a limited capacity for experiencing pleasure, expressing emotions, and desiring intimacy. While therapy might gently encourage a broader range of emotional awareness and expression if the client desires it and feels safe, trying to promote "greater emotional intensity" directly would likely be counterproductive, increasing anxiety and resistance rather than fostering genuine change or comfort.

### Classificazione dei Disturbi di Personalità (DSM-5)

I disturbi di personalità sono raggruppati in tre cluster principali basati su somiglianze sintomatologiche:

*   **Cluster A (Eccentrico/Paranoide):** Caratterizzato da comportamenti strani o eccentrici. Include i disturbi Paranoide, Schizoide e Schizotipico. Il tema dominante è la sfiducia e il distacco sociale.
*   **Cluster B (Drammatico/Impulsivo):** Caratterizzato da comportamenti drammatici, emotivi o irregolari. Include i disturbi Antisociale, Borderline, Istrionico e Narcisistico. Il tema dominante è l'instabilità emotiva e la scarsa regolazione degli impulsi.
*   **Cluster C (Ansioso/Pauroso):** Caratterizzato da comportamenti ansiosi o timorosi. Include i disturbi Evitante, Dipendente e Ossessivo-Compulsivo di personalità. Il tema dominante è l'ansia e l'inibizione.

### Principi Generali del Trattamento Psicoterapeutico

Indipendentemente dal cluster specifico, la psicoterapia per i disturbi di personalità segue alcuni principi cardine:

1.  **Alleanza Terapeutica:** La costruzione di un legame di fiducia è fondamentale, poiché le difficoltà relazionali sono spesso il nucleo del disturbo.
2.  **Stabilità del Setting:** Un quadro terapeutico chiaro e costante (orari, frequenza, regole) fornisce la sicurezza necessaria per esplorare vissuti profondi.
3.  **Gestione del Controtransfert:** Il terapeuta deve monitorare attentamente le proprie reazioni emotive, spesso intense, suscitate dal paziente per evitare agiti non terapeutici.
4.  **Natura a Lungo Termine:** Trattandosi di tratti strutturati della personalità, il cambiamento richiede tempi prolungati e interventi costanti.

### Impatto sul Funzionamento

I disturbi di personalità compromettono significativamente il funzionamento sociale e lavorativo, portando a isolamento, conflitti interpersonali cronici e instabilità professionale. Un approccio strutturato è essenziale per ridurre la sofferenza soggettiva e migliorare l'adattamento globale dell'individuo.

## Trattamento del Cluster A (Eccentrico/Paranoide)

### Subtask:
Analizzare le strategie terapeutiche specifiche per i disturbi Paranoide, Schizoide e Schizotipico.


### Strategie Terapeutiche per il Cluster A (Eccentrico/Paranoide)

Il trattamento dei disturbi del Cluster A (Paranoide, Schizoide, Schizotipico) presenta sfide uniche a causa della natura stessa di questi disturbi, spesso caratterizzati da diffidenza e distacco sociale.

#### 1. Obiettivo Primario della Psicoterapia
L'obiettivo fondamentale per l'intero Cluster A è la creazione di un'**alleanza terapeutica basata sulla fiducia**. È cruciale che il terapeuta mantenga una posizione di **rispetto per la distanza relazionale** del paziente. Tentativi prematuri di forzare l'intimità o l'esplorazione emotiva profonda possono essere percepiti come minacciosi, portando al ritiro o all'abbandono della terapia.

#### 2. Approccio per il Disturbo Paranoide
*   **Neutralità e Trasparenza:** Il terapeuta deve adottare uno stile professionale, onesto e non eccessivamente caloroso.
*   **Evitare il Sospetto:** Ogni intervento, inclusi i cambiamenti di orario o le spiegazioni diagnostiche, deve essere comunicato con estrema chiarezza e logica per evitare di alimentare le interpretazioni malevole o paranoiche del paziente.

#### 3. Strategie per il Disturbo Schizoide
*   **Connessioni Sicure:** L'obiettivo non è trasformare il paziente in un estroverso, ma favorire connessioni sociali che siano **sicure e gestibili**.
*   **Rispetto dell'Isolamento:** Le strategie sono orientate a ridurre l'isolamento distruttivo sviluppando abilità sociali pratiche, senza forzare una condivisione emotiva che il paziente potrebbe trovare invadente o priva di significato.

#### 4. Approccio per il Disturbo Schizotipico
*   **Test di Realtà:** La terapia si concentra spesso sulla gestione del pensiero magico e delle idee di riferimento, aiutando il paziente a distinguere tra le proprie percezioni insolite e la realtà oggettiva.
*   **Sviluppo di Abilità Sociali:** Si lavora sulla riduzione dell'ansia sociale e sul miglioramento della comunicazione interpersonale per facilitare l'adattamento quotidiano.

#### 5. Ruolo della Farmacoterapia di Supporto
Sebbene la psicoterapia sia il trattamento d'elezione, la farmacoterapia gioca un ruolo di supporto critico:
*   **Antipsicotici a basso dosaggio:** Possono essere utilizzati per gestire sintomi "quasi-psicotici" (come sospettosità estrema o distorsioni cognitive) o ansia grave.
*   **Limiti:** I farmaci non rappresentano la cura primaria per i tratti di personalità, ma servono a stabilizzare il paziente affinché possa beneficiare maggiormente degli interventi psicosociali.

## Trattamento del Cluster B (Drammatico/Impulsivo)

### Subtask:
Approfondire le strategie terapeutiche per i disturbi Borderline, Narcisistico, Antisociale e Istrionico, focalizzandosi sulla regolazione emotiva e la gestione dell'impulsività.


## Strategie Terapeutiche per il Cluster B (Drammatico/Impulsivo)

### 1. Disturbo Borderline di Personalità (DBP)
Il trattamento del DBP si concentra sulla stabilizzazione affettiva e la riduzione dei comportamenti autolesivi.
*   **Terapia Dialettico-Comportamentale (DBT):** Considerata il gold standard, combina tecniche cognitive con la filosofia zen. Si basa sulla dialettica tra accettazione e cambiamento, insegnando abilità di *mindfulness*, efficacia interpersonale, regolazione emotiva e tolleranza dell'angoscia.
*   **Psicoterapia Focalizzata sul Transfert (TFP):** Un approccio psicodinamico che utilizza la relazione tra paziente e terapeuta per integrare le rappresentazioni scisse di sé e degli altri, migliorando il controllo degli impulsi e la coerenza dell'identità.

### 2. Disturbo Narcisistico di Personalità (DNP)
La terapia per il DNP è spesso ostacolata dalla resistenza del paziente a mostrare vulnerabilità.
*   **Gestione dell'Autostima:** Il focus è sull'esplorazione della fragilità sottostante al senso di grandiosità. Il terapeuta deve bilanciare la validazione del bisogno di ammirazione con la messa in discussione delle distorsioni cognitive.
*   **Alleanza Terapeutica:** È fondamentale gestire le reazioni di rabbia narcisistica e le svalutazioni del terapeuta per mantenere la continuità del trattamento.

### 3. Disturbo Antisociale di Personalità (DAP)
La psicoterapia tradizionale spesso fallisce a causa della mancanza di rimorso o desiderio di cambiamento del paziente.
*   **Responsabilità e Conseguenze:** Il trattamento si sposta dal 'perché' al 'cosa'. Si punta a far comprendere al paziente che i comportamenti prosociali sono più vantaggiosi a lungo termine rispetto a quelli manipolatori o aggressivi.
*   **Setting Strutturati:** Spesso richiede interventi in contesti controllati o programmi di gruppo focalizzati sulla correzione del pensiero distorto e sulla gestione della rabbia.

### 4. Disturbo Istrionico di Personalità (DIP)
L'obiettivo è spostare l'attenzione dalla ricerca di approvazione esterna alla comprensione interna.
*   **Comunicazione Autentica:** Il terapeuta aiuta il paziente a identificare i sentimenti reali dietro le manifestazioni teatrali, incoraggiando un linguaggio meno impressionistico e più analitico.
*   **Limiti e Confini:** La terapia lavora sulla tolleranza alla frustrazione quando il paziente non è al centro dell'attenzione.

### 5. Schema Therapy e Modulazione dell'Impulsività
La **Schema Therapy** è particolarmente efficace per l'intero Cluster B poiché:
*   Identifica i **'Mode'** (stati emotivi transitori come il 'Bambino Abbandonato' o il 'Genitore Punitivo') che guidano l'instabilità.
*   Utilizza il **re-parenting limitato** per fornire al paziente la stabilità emotiva mancata nell'infanzia.
*   Aiuta a indebolire gli schemi disadattivi precoci, sostituendo le reazioni impulsive con risposte adulte e autoriflessive.

## Trattamento del Cluster C (Ansioso/Pauroso)

### Subtask:
Esplorare le strategie terapeutiche specifiche per i disturbi Evitante, Dipendente e Ossessivo-Compulsivo di Personalità.


### Strategie Terapeutiche per il Cluster C (Ansioso/Pauroso)

#### 1. Obiettivi Principali della Psicoterapia
Il focus primario del trattamento per i disturbi del Cluster C è la **riduzione dell'ansia e dell'inibizione**, parallelamente allo **sviluppo dell'autonomia e dell'efficacia personale**. L'obiettivo è aiutare il paziente a riconoscere e tollerare l'incertezza, diminuendo la dipendenza da rigide routine o dal supporto esterno per connettersi con i propri desideri autentici.

#### 2. Disturbo Evitante di Personalità
L'approccio si concentra sul superamento del senso di inadeguatezza e del timore del rifiuto:
*   **Esposizione Graduale:** Incoraggiare il paziente ad affrontare situazioni sociali temute per disconfermare le aspettative di critica.
*   **Ristrutturazione Cognitiva:** Sfidare i pensieri automatici legati all'inferiorità e alla convinzione di essere socialmente inetti.
*   **Social Skills Training:** Migliorare le abilità relazionali per aumentare la fiducia in contesti interpersonali.

#### 3. Disturbo Dipendente di Personalità
La sfida centrale è gestire il bisogno eccessivo di essere accuditi senza creare una nuova dipendenza dal terapeuta:
*   **Promozione dell'Autonomia:** Incoraggiare piccoli passi verso il decision-making indipendente.
*   **Gestione del Transfert:** Monitorare la tendenza del paziente a delegare la responsabilità delle proprie scelte al terapeuta.
*   **Assertività:** Sviluppare la capacità di esprimere disaccordo e bisogni personali senza il timore di abbandono.

#### 4. Disturbo Ossessivo-Compulsivo di Personalità (OCPD)
Il trattamento mira a smussare il perfezionismo e la rigidità:
*   **Flessibilità Cognitiva:** Aiutare il paziente a comprendere che esistono molteplici modi "corretti" di agire, riducendo il pensiero dicotomico (tutto o nulla).
*   **Accettazione dell'Imperfezione:** Lavorare sul valore del 'buono abbastanza' rispetto all'eccellenza assoluta.
*   **Connessione Emotiva:** Spostare l'attenzione dalle 'regole' ai 'bisogni', aiutando il paziente a riconoscere le emozioni che il controllo rigido tenta di soffocare.

#### 5. Efficacia dei Modelli Terapeutici
*   **Terapia Cognitivo-Comportamentale (CBT):** Risulta altamente efficace nel Cluster C per modificare i pattern di pensiero disfunzionali e promuovere cambiamenti comportamentali attraverso l'esposizione e il problem-solving.
*   **Psicoterapia Psicodinamica a Breve Termine:** Utile per esplorare le radici dell'ansia e dell'inibizione, favorendo un'integrazione emotiva che riduce la necessità di meccanismi di difesa basati sull'evitamento o sulla sottomissione.

### Approcci Farmacoterapeutici di Supporto

#### 1. Il Ruolo della Farmacoterapia nei Disturbi di Personalità
È fondamentale chiarire che la farmacoterapia **non agisce sulla struttura sottostante della personalità**, la quale rimane l'obiettivo primario dell'intervento psicoterapeutico. Il ruolo dei farmaci è prettamente di **supporto**: essi servono a stabilizzare i cosiddetti **sintomi 'bersaglio'** (es. impulsività estrema, ansia paralizzante, disregolazione emotiva). Riducendo l'intensità di questi sintomi, il farmaco crea una 'finestra terapeutica' che rende il paziente più accessibile e capace di impegnarsi efficacemente nel lavoro psicologico.

#### 2. Classificazione dei Farmaci e Obiettivi Specifici
La scelta farmacologica si basa sulla sintomatologia prevalente:
*   **SSRI (Inibitori Selettivi della Ricaptazione della Serotonina):** Utilizzati principalmente per gestire la **disregolazione affettiva**, l'irritabilità e l'**impulsività cognitiva** (rimuginio, ossessioni).
*   **Stabilizzatori dell'Umore (es. Litio, Anticonvulsivanti come Valproato o Lamotrigina):** Indicati per l'**instabilità comportamentale**, i rapidi cambiamenti d'umore e le esplosioni di **rabbia**.
*   **Antipsicotici Atipici a basso dosaggio (es. Quetiapina, Olanzapina, Aripiprazolo):** Impiegati per il controllo di **sintomi cognitivo-percettivi** (ideazione paranoide transitoria, dissociazione), ansia grave o marcata disorganizzazione del pensiero.

#### 3. Monitoraggio, Aderenza e Sicurezza
Il monitoraggio deve essere rigoroso e costante per diversi motivi:
*   **Effetti Collaterali:** Valutare l'impatto metabolico o sedativo che potrebbe ostacolare la partecipazione alla terapia.
*   **Aderenza:** Nei pazienti con tendenze manipolatorie o tratti Borderline, l'aderenza può essere incostante (uso improprio del farmaco come gesto autolesivo o 'test' verso il medico).
*   **Sicurezza:** È cruciale limitare la prescrizione di farmaci letali in overdose (come i triciclici) in pazienti con alto rischio di impulsività o gesti anticonservativi.

'farmaco per il Disturbo Borderline'), la scelta clinica deve essere guidata dalla **dimensione sintomatologica prevalente**.
