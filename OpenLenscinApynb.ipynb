### OpenLens AI Project Structure

Here is a quick overview of the key directories and files in the `openlens_ai/` package:

*   `agents/`: Contains specialized agent logic (Coding, Analysis, LaTeX Writing, Literature Review, and Supervision).
*   `prompts/`: LLM prompt templates used by the agents.
*   `tools/`: Custom tools and utility scripts.
*   `build_graph.py`: The core logic for constructing the LangGraph workflow.
*   `frontend.py`: The Streamlit-based user interface.
*   `state.py`: Management of the application state and shared memory.

```python
import toml
import os
from loguru import logger
from tavily import TavilyClient
import google.generativeai as genai

class MedicalExpertAgent:
    def __init__(self, config_path="openlens-ai/config.toml"):
        self.config = toml.load(config_path)
        self.chat_model_name = self.config['llm']['chat']['model']
        self.tavily_key = self.config['tools']['tavily_api_key']
        self.language = "ita"

        # Inizializziamo il client Tavily
        if self.tavily_key and "INSERISCI" not in self.tavily_key:
            self.tavily = TavilyClient(api_key=self.tavily_key)
        else:
            self.tavily = None
            logger.warning("Tavily API Key non configurata nel file toml.")

    def search_latest_therapy(self, diagnosis, biomarkers):
        """
        Esegue una ricerca reale tramite Tavily e genera una sintesi clinica.
        """
        query = f"Latest clinical guidelines and targeted therapies for {diagnosis} with {biomarkers} 2024 2025"

        search_context = ""
        if self.tavily:
            logger.info(f"Ricerca Tavily in corso per: {query}")
            search_result = self.tavily.search(query=query, search_depth="advanced", max_results=5)
            search_context = "\n".join([f"Fonte: {r['url']}\nContenuto: {r['content']}" for r in search_result['results']])

        prompt = f"""
        Ruolo: Oncologo Specialista Senior.
        Task: Fornire raccomandazioni terapeutiche basate su evidenze.

        Diagnosi: {diagnosis}
        Biomarcatori: {biomarkers}

        Contesto di ricerca trovato:\n{search_context if search_context else 'Nessun dato esterno trovato, usa la tua conoscenza interna.'}

        Istruzioni:
        1. Sintetizza le opzioni terapeutiche (chirurgia, TKI, adiuvante).
        2. Cita studi specifici se presenti nel contesto.
        3. Rispondi in lingua: {self.language}.
        """

        logger.info("Generazione risposta medica in corso...")
        return f"[Simulazione Output LLM basata su Tavily]\nAnalisi per {diagnosis} ({biomarkers}):\nL'integrazione Tavily è attiva. L'agente cercherà i dati e li elaborerà con {self.chat_model_name}."

# Re-inizializzazione
med_agent = MedicalExpertAgent()
print("Agente Medico con integrazione Tavily configurato.")
```

### 🏥 Analisi del Caso Clinico: Infezione da Citomegalovirus (CMV)

Il quadro descritto (ittero progressivo, petecchie, ipotonia, letargia) in una neonata di una settimana con PCR positiva per CMV è altamente suggestivo di **infezione congenita da CMV**.

Analizziamo le opzioni per identificare l'elemento **NON tipico**:

1.  **Iperbilirubinemia diretta**: **TIPICO**. Il CMV causa epatite neonatale con colestasi, portando a un aumento della bilirubina coniugata (diretta).
2.  **Aumento degli indici epatici**: **TIPICO**. Il danno citopatico alle cellule epatiche causa l'innalzamento di AST e ALT.
3.  **Deficit di crescita**: **TIPICO**. I neonati con CMV sintomatico sono spesso piccoli per l'età gestazionale (SGA) e presentano scarso accrescimento post-natale.
4.  **Riduzione delle concentrazioni di vitamina K**: **NON TIPICO/INDIRETTO**. Sebbene una colestasi grave possa ridurre l'assorbimento delle vitamine liposolubili, il deficit di Vitamina K non è un segno cardine o specifico del CMV rispetto ad altre patologie della coagulazione o emorragiche del neonato.
5.  **Riduzione dei livelli di Aptoglobina**: **ELEMENTO NON TIPICO**. L'aptoglobina si riduce nelle **anemie emolitiche**. Sebbene il CMV possa causare anemia, questa è solitamente di tipo iporigenerativo (soppressione midollare) o legata alla malattia cronica, non un'emolisi intravascolare massiva tipica dei quadri in cui l'aptoglobina crolla.

**Risposta corretta al quesito:**
In ambito accademico, l'elemento che meno si sposa con il quadro di epatite/piastrinopenia da CMV è la **riduzione dei livelli di Aptoglobina** (o l'aptoglobina stessa come marker, in quanto l'ittero del CMV è ostruttivo/epatico, non emolitico).

### 🔍 Approfondimento: Aptoglobina e Ittero nel CMV

Per capire perché la riduzione dell'aptoglobina non è tipica, dobbiamo distinguere tra **Ittero Emolitico** e **Ittero Epatocellulare/Colestatico**.

#### 1. Cos'è l'Aptoglobina?
L'aptoglobina è una proteina prodotta dal fegato che ha un compito specifico: legarsi all'emoglobina libera nel plasma. Quando i globuli rossi si rompono all'interno dei vasi sanguigni (**emolisi intravascolare**), l'emoglobina fuoriesce; l'aptoglobina la cattura per evitare che danneggi i reni. In questo processo, l'aptoglobina viene consumata e i suoi livelli nel sangue crollano.

#### 2. Cosa succede nel CMV?
L'infezione da CMV causa un ittero di tipo **epatocellulare**:
*   Il virus attacca direttamente gli epatociti (cellule del fegato).
*   Il danno causa un'infiammazione che impedisce il normale deflusso della bile (colestasi).
*   La bilirubina che aumenta è quella **diretta (coniugata)**, perché il fegato riesce a trasformarla ma non riesce a espellerla.
*   **Non c'è una distruzione massiva di globuli rossi.** Di conseguenza, l'aptoglobina non ha motivo di ridursi, poiché non c'è emoglobina libera da raccogliere.

#### 3. Perché le altre opzioni erano tipiche?
*   **Vitamina K**: La colestasi (blocco della bile) impedisce l'assorbimento dei grassi e delle vitamine liposolubili (A, D, E, K). Anche se non è il segno principale, è una conseguenza fisiopatologica possibile del danno epatico.
*   **Deficit di crescita**: Il virus sottrae nutrienti e danneggia diversi organi durante lo sviluppo fetale, portando a neonati SGA (Small for Gestational Age).

**In sintesi:** Se trovi l'aptoglobina bassa, devi sospettare un'anemia emolitica (come l'incompatibilità AB0 o Rh), non un'infezione virale come il CMV.

```python
!pip install biopython

import os
from Bio import Entrez
import pandas as pd

class PubMedResearchAgent:
    def __init__(self, email="your.email@example.com"):
        # Entrez richiede un'email per identificare chi effettua le chiamate
        Entrez.email = email

    def search_and_summarize(self, query, max_results=5):
        """
        Cerca su PubMed, estrae abstract e crea un riassunto con link.
        """
        print(f"🔍 Ricerca PubMed in corso per: {query}...")

        # 1. Ricerca degli ID
        handle = Entrez.esearch(db="pubmed", term=query, retmax=max_results)
        record = Entrez.read(handle)
        handle.close()

        id_list = record["IdList"]
        if not id_list:
            return "Nessun risultato trovato."

        # 2. Recupero dettagli
        handle = Entrez.efetch(db="pubmed", id=",".join(id_list), rettype="medline", retmode="text")
        from Bio import Medline
        records = Medline.parse(handle)

        results = []
        for record in records:
            title = record.get("TI", "N/A")
            authors = ", ".join(record.get("AU", []))
            source = record.get("SO", "N/A")
            pmid = record.get("PMID", "N/A")
            abstract = record.get("AB", "Abstract non disponibile.")
            link = f"https://pubmed.ncbi.nlm.nih.gov/{pmid}/"

            results.append({
                "Titolo": title,
                "Autori": authors,
                "Fonte": source,
                "PMID": pmid,
                "Link": link,
                "Riassunto": abstract[:300] + "..." # Taglio per brevità
            })

        return pd.DataFrame(results)

# Inizializzazione e Test dell'Agente PubMed
pubmed_agent = PubMedResearchAgent()

# Esempio: Ricerca su immunoterapia e NSCLC
query_medica = "NSCLC immunotherapy targeted therapy 2024"
df_bibliografia = pubmed_agent.search_and_summarize(query_medica)

display(df_bibliografia[["Titolo", "Autori", "Link", "Riassunto"]])
```

### 🏥 Gestione Post-TC Negativa nel Paziente Anticoagulato

Anche se la TC iniziale eseguita in emergenza risulta negativa, il paziente in terapia con anticoagulanti (Warfarin, NOAC/DOAC) non può essere dimesso immediatamente con la stessa leggerezza di un paziente non scoagulato.

#### 1. Rischio di Emorragia Ritardata
Esiste una finestra temporale (solitamente nelle prime 24 ore) in cui un piccolo sanguinamento, inizialmente non visibile alla TC, può espandersi a causa del difetto della coagulazione.

#### 2. Protocolli di Osservazione
Le linee guida (come quelle della *Società Italiana di Medicina d'Emergenza-Urgenza* o il protocollo *NICE*) suggeriscono due approcci principali:
*   **Osservazione Clinica:** Ricovero o osservazione in OBI (Osservazione Breve Intensiva) per almeno **24 ore** con monitoraggio neurologico seriato (GCS, pupille, deficit focali).
*   **Ripetizione della TC:** Alcuni centri eseguono una **seconda TC di controllo a 24 ore** prima della dimissione, specialmente se il trauma è stato ad alta energia o se l'INR (nel caso del warfarin) è sovraterapeutico (>3.0).

#### 3. Istruzioni alla Dimissione (Safety Netting)
Alla dimissione, il paziente e i familiari devono ricevere istruzioni scritte per tornare immediatamente in PS se compaiono:
*   Sonnolenza eccessiva o difficoltà al risveglio.
*   Confusione o irritabilità.
*   Vomito persistente.
*   Cefalea che peggiora nonostante gli analgesici comuni.
*   Debolezza agli arti o problemi di coordinazione.

**Nota:** È fondamentale rivalutare la terapia anticoagulante (se sospenderla temporaneamente o proseguirla) in base al rischio trombotico individuale del paziente.

### 🏥 Protocollo di Osservazione 24h in OBI

L'obiettivo principale dell'osservazione in OBI per il trauma cranico minore nel paziente anticoagulato è l'individuazione precoce di segni di **ipertensione endocranica** o **focalità neurologica**.

#### 1. Monitoraggio Neurologico (GCS e Pupille)
Deve essere eseguito ogni ora per le prime 6 ore, poi ogni 2-4 ore se il quadro rimane stabile:
*   **Glasgow Coma Scale (GCS):** Un calo di anche solo 1 punto (specialmente nella risposta motoria) deve allertare il medico.
*   **Esame Pupillare:** Valutazione di diametro, simmetria (anisocoria) e reattività alla luce. Una pupilla 'pigra' o dilatata può indicare una compressione del III nervo cranico per erniazione uncale.

#### 2. Parametri Vitali (Triade di Cushing)
Il monitoraggio pressorio è critico. Bisogna vigilare sulla comparsa della **Triade di Cushing**, segno tardivo di ipertensione endocranica:
1.  **Ipertensione arteriosa** (con aumento della pressione differenziale).
2.  **Bradicardia**.
3.  **Alterazioni del ritmo respiratorio**.

#### 3. Gestione dei Sintomi
*   **Cefalea:** Deve essere valutata la risposta agli analgesici. Una cefalea che progredisce nonostante il paracetamolo è un segnale di allarme.
*   **Vomito:** Se diventa 'a getto' o incoercibile, indica un aumento della pressione intracranica.

#### 4. Criteri per la Ripetizione della TC (TC 'Second Look')
La TC va ripetuta immediatamente se:
*   Il GCS scende di $\geq 1$ punto.
*   Compaiono nuovi deficit focali o convulsioni.
*   La cefalea diventa insopportabile.
*   In assenza di sintomi, molti protocolli prevedono comunque una TC di controllo a 24h se l'**INR è > 3.0** o in caso di trauma ad alta energia.
