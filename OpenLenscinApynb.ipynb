### 🔑 Come ottenere il tuo Ngrok Auth Token

1.  **Accedi**: Vai su [dashboard.ngrok.com](https://dashboard.ngrok.com/) ed esegui il login (o registrati gratuitamente).
2.  **Trova il Token**: Nel menu a sinistra, clicca sulla sezione **'Your Authtoken'** (sotto la voce *Getting Started*).
3.  **Copia**: Vedrai una stringa di caratteri in un riquadro grigio. Clicca sul pulsante **'Copy'**.
4.  **Incolla**: Torna nella cella di codice sopra, cancella `INSERISCI_IL_TUO_TOKEN_QUI` e incolla il tuo codice mantenendo le virgolette.
5.  **Esegui**: Avvia la cella premendo il tasto Play.

### OpenLens AI Project Structure

Here is a quick overview of the key directories and files in the `openlens_ai/` package:

*   `agents/`: Contains specialized agent logic (Coding, Analysis, LaTeX Writing, Literature Review, and Supervision).
*   `prompts/`: LLM prompt templates used by the agents.
*   `tools/`: Custom tools and utility scripts.
*   `build_graph.py`: The core logic for constructing the LangGraph workflow.
*   `frontend.py`: The Streamlit-based user interface.
*   `state.py`: Management of the application state and shared memory.
