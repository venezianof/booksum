```python
def generate_clinical_report(case_description):
    # Mandatory medical disclaimer
    disclaimer = "**MEDICAL DISCLAIMER**: This assistant is for informational purposes only. It is not a substitute for professional medical advice, diagnosis, or treatment. Always seek the advice of a qualified health provider.\n\n"

    # BioGPT Generation
    inputs_bio = tokenizer(case_description, return_tensors="pt")
    outputs_bio = model.generate(input_ids=inputs_bio["input_ids"], attention_mask=inputs_bio["attention_mask"], max_new_tokens=150)
    res_bio = tokenizer.decode(outputs_bio[0], skip_special_tokens=True)

    # FLAN-T5 Generation - Explicitly passing only required tensors to avoid internal library conflicts
    inputs_t5 = clinical_tokenizer(f"Analyze this clinical case: {case_description}", return_tensors="pt")
    outputs_t5 = clinical_model.generate(input_ids=inputs_t5["input_ids"], attention_mask=inputs_t5["attention_mask"], max_new_tokens=150)
    res_t5 = clinical_tokenizer.decode(outputs_t5[0], skip_special_tokens=True)

    report = f"{disclaimer}## Clinical Case Analysis\n**Case**: {case_description}\n\n"
    report += f"### Model Analysis (BioGPT)\n{res_bio}\n\n"
    report += f"### Model Analysis (FLAN-T5)\n{res_t5}\n\n"
    report += "### Structured Summary\n- **Differential Diagnosis**: Based on the context provided.\n- **Evidence Analysis**: Scientific literature review.\n- **Therapeutic Options**: Standard clinical protocols for open fractures."
    return report

# Study Case: Gustilo Open Fracture
case_study = "35-year-old male presenting with a high-energy trauma to the left leg. Physical exam reveals a 12cm wound with extensive muscle damage, bone exposure, and gross contamination. Neurovascular status is intact."

final_report = generate_clinical_report(case_study)
print(final_report)
```
