### ūüõ†ÔłŹ Monitoraggio Semplificato
Esegui questa cella per vedere lo stato dei processi e gli eventuali errori in tempo reale.

````python
import os
import time

def check_logs():
    print("--- STATO PROCESSI ---")
    !ps -ef | grep -E 'node|npm' | grep -v grep

    log_files = [
        '/content/dev_stdout.log',
        '/content/dev_stderr.log',
        '/content/worker_stdout.log',
        '/content/worker_stderr.log'
    ]

    for log in log_files:
        print(f"\n--- Analisi file: {log} ---")
        if os.path.exists(log):
            size = os.path.getsize(log)
            if size > 0:
                with open(log, 'r') as f:
                    # Legge le ultime 20 righe
                    lines = f.readlines()
                    for line in lines[-20:]:
                        print(line.strip())
            else:
                print("Il file esiste ma √® VUOTO (il processo non ha scritto nulla).")
        else:
            print("File non trovato. Il processo non √® partito o non ha permessi di scrittura.")

check_logs()
````
