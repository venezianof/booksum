**Reasoning**:
I will define the advanced medical expert system prompt and implement the `analizza_caso_clinico` function to structure clinical analysis according to the specified requirements.



```python
# 1. Definizione del System Prompt avanzato
SYSTEM_PROMPT_MEDICO = """Sei un Medico Specialista in Medicina Interna e Farmacologia Clinica con decenni di esperienza.
Il tuo compito  analizzare casi clinici complessi, spesso caratterizzati da comorbidit  e polifarmacoterapia.

ISTRUZIONI DI STRUTTURA:
Rispondi SEMPRE seguendo rigorosamente questo schema:
1. **Analisi dei Sintomi**: Valutazione fisiopatologica dei segni e sintomi riferiti.
2. **Diagnosi Differenziale**: Elenco delle ipotesi diagnostiche plausibili con relativa giustificazione.
3. **Inquadramento Clinico**: Sintesi della condizione basata sulle linee guida internazionali (es. ESC/ESH, ADA, GOLD).
4. **Suggerimenti Terapeutici**: Proposta di gestione farmacologica, con attenzione critica alle interazioni farmacologiche e aggiustamenti per comorbidit .

LINEE GUIDA:
- Mantieni un tono professionale, scientifico e conciso.
- Cita le linee guida di riferimento ove pertinente.
- Evidenzia potenziali rischi di tossicit  o controindicazioni.
- Rispondi esclusivamente in lingua italiana."""

def analizza_caso_clinico(descrizione_caso, modello=llm):
    """
    Funzione per l'analisi di casi clinici complessi utilizzando parametri deterministici.
    """
    # 4. Composizione del prompt con tag Llama-3.2
    full_prompt = f"<|start_header_id|>system<|end_header_id|>\n\n{SYSTEM_PROMPT_MEDICO}<|eot_id|><|start_header_id|>user<|end_header_id|>\n\nAnalizza il seguente caso clinico: {descrizione_caso}<|eot_id|><|start_header_id|>assistant<|end_header_id|>"

    # 5. Configurazione parametri deterministici
    params = {
        "max_tokens": 1024,
        "temperature": 0.1,
        "top_p": 0.9,
        "echo": False
    }

    response = modello(
        full_prompt,
        **params
    )

    return response['choices'][0]['text'].strip()

print("Sistema di prompt e funzione 'analizza_caso_clinico' configurati.")
```
