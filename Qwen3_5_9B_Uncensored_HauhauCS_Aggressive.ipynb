```python
!pip install llama-cpp-python huggingface_hub

from llama_cpp import Llama
from huggingface_hub import hf_hub_download

print("1. Scaricamento del modello in corso (circa 5GB)... attendere.")
# Repository verificato e pubblico per Llama 3.1 8B Instruct GGUF
model_path = hf_hub_download(
    repo_id="QuantFactory/Meta-Llama-3.1-8B-Instruct-GGUF",
    filename="Meta-Llama-3.1-8B-Instruct.Q4_K_M.gguf"
)

print("2. Caricamento del modello nella memoria della GPU...")
llm = Llama(
    model_path=model_path,
    n_ctx=2048,
    n_gpu_layers=-1
)

def chiedi_all_agente(domanda):
    system_msg = "Sei un esperto medico. Fornisci risposte tecniche, precise e basate su evidenze scientifiche."
    res = llm.create_chat_completion(
        messages=[{"role": "system", "content": system_msg}, {"role": "user", "content": domanda}],
        temperature=0.2
    )
    return res['choices'][0]['message']['content']

print("\n--- AGENTE PRONTO PER L'USO! ---")
```

```python
# SCRIVI QUI LA TUA DOMANDA
quesito = "Quali sono i rischi della terapia con ACE-inibitori in gravidanza?"

risposta = chiedi_all_agente(quesito)
print(risposta)
```

```python
# !pip install llama-cpp-python

from llama_cpp import Llama

llm = Llama.from_pretrained(
	repo_id="HauhauCS/Qwen3.5-9B-Uncensored-HauhauCS-Aggressive",
	filename="Qwen3.5-9B-Uncensored-HauhauCS-Aggressive-BF16.gguf",
)
```

```python
# Definizione del prompt di sistema per l'agente medico
system_prompt = """Sei un assistente AI specializzato in ricerca medica avanzata e supporto alla terapia per medici professionisti.
Il tuo obiettivo è fornire analisi basate su evidenze scientifiche, riassumere studi clinici e suggerire protocolli terapeutici aggiornati.

Regole:
1. Fornisci riferimenti a meccanismi fisiopatologici.
2. Includi avvertenze sulle interazioni farmacologiche.
3. Mantieni un tono tecnico e professionale.
4. Ricorda sempre che i tuoi suggerimenti devono essere validati da un medico clinico."""

def ricerca_medica(quesito):
    messages = [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": quesito}
    ]

    response = llm.create_chat_completion(
        messages=messages,
        temperature=0.2, # Bassa temperatura per maggiore precisione
        max_tokens=1024
    )
    return response['choices'][0]['message']['content']

# Esempio di utilizzo
quesito_esempio = "Analisi dei benefici della metformina nel trattamento dell'ovaio policistico (PCOS) e potenziali effetti collaterali a lungo termine."
risultato = ricerca_medica(quesito_esempio)
print(risultato)
```
