```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Dati reali DAPA-HF per la visualizzazione
labels = ['Endpoint Primario Raggiunto', 'Endpoint Non Raggiunto']
sizes_dapa = [16.3, 83.7]  # Percentuale nel gruppo Dapagliflozin
sizes_placebo = [21.2, 78.8] # Percentuale nel gruppo Placebo

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Grafico a torta: Incidenza Eventi nel gruppo Dapagliflozin
ax1.pie(sizes_dapa, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
ax1.set_title('Incidenza Endpoint Primario\n(Gruppo Dapagliflozin)')

# Grafico a barre: Hazard Ratio (HR) con Intervallo di Confidenza
hr_data = {'Trial': 'DAPA-HF', 'HR': 0.74, 'Lower_CI': 0.65, 'Upper_CI': 0.85}
ax2.errorbar(x=hr_data['HR'], y=0, xerr=[[hr_data['HR']-hr_data['Lower_CI']], [hr_data['Upper_CI']-hr_data['HR']]], fmt='o', color='black', capsize=10, markersize=10)
ax2.axvline(x=1.0, color='red', linestyle='--') # Linea di non efficacia
ax2.set_xlim(0.5, 1.2)
ax2.set_yticks([])
ax2.set_xlabel('Hazard Ratio (IC 95%)')
ax2.set_title('Efficacia del Trattamento (HR < 1 favorisce Dapagliflozin)')

plt.tight_layout()
plt.show()
```

```python
def med_research_agent(query):
    system_prompt = """Sei un assistente IA specializzato in ricerca medica avanzata e farmacoterapia per medici specialisti.
    Il tuo obiettivo è fornire analisi basate sull'evidenza (EBM) con un focus rigoroso sui trial clinici.

    Regole Mandatorie:
    1. Cita esplicitamente i trial clinici cardine (es. EMPEROR, DAPA-CKD, LEADER) e l'anno di pubblicazione se disponibile.
    2. Usa terminologia medica di alto livello (es. farmacocinetica, endpoint compositi, HR, IC 95%).
    3. Struttura: Meccanismo Molecolare, Analisi Trial Clinici (Evidenze), Considerazioni Cliniche e Linee Guida (ESC/EASD/KDIGO).
    4. Se i dati sono contrastanti tra diversi trial, evidenzia la discrepanza.
    5. Mantieni un tono professionale, sintetico e analitico."""

    response = llm.create_chat_completion(
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": query}
        ],
        max_tokens=2048,
        temperature=0.2,
        top_p=0.95,
        repeat_penalty=1.2
    )
    return response['choices'][0]['message']['content']

print("System Prompt ottimizzato per il richiamo dei trial clinici.")
```

```python
# Esempio di utilizzo per un medico specialista
quesito = "Analisi dell'efficacia degli inibitori SGLT2 nello scompenso cardiaco a frazione di eiezione preservata (HFpEF) basata sui trial EMPEROR-Preserved e DELIVER."

risultato = med_research_agent(quesito)
print(risultato)
```
