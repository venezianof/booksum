```python
import matplotlib.pyplot as plt
import pandas as pd

# Dati esemplificativi per NOAC e riduzione dello stroke, basati su meta-analisi di trial clinici chiave
# Valori HR e IC 95% sono fittizi ma rappresentativi delle tendenze osservate nei trial RE-LY, ARISTOTLE, ENGAGE AF-TIMI 48, ROCKET AF
data = {
    'NOAC': ['Apixaban', 'Dabigatran', 'Edoxaban', 'Rivaroxaban'],
    'HR': [0.79, 0.65, 0.79, 0.82], # Hazard Ratio vs Warfarin per l'endpoint primario (stroke/embolia sistemica)
    'Lower_CI': [0.69, 0.52, 0.68, 0.70],
    'Upper_CI': [0.90, 0.81, 0.92, 0.96]
}
df_noac = pd.DataFrame(data)

plt.figure(figsize=(10, 6))

# Creazione del Forest Plot
for i, row in df_noac.iterrows():
    plt.errorbar(x=row['HR'], y=i,
                 xerr=[[row['HR'] - row['Lower_CI']], [row['Upper_CI'] - row['HR']]],
                 fmt='o', color='darkgreen', capsize=8, markersize=10,
                 label=row['NOAC'] if i == 0 else "",
                 zorder=3) # Assicura che i punti siano sopra la griglia
    plt.text(row['Upper_CI'] + 0.02, i, f"HR = {row['HR']:.2f} ({row['Lower_CI']:.2f}-{row['Upper_CI']:.2f})",
             va='center', ha='left', fontsize=10)

plt.axvline(x=1.0, color='red', linestyle='--', label='Linea di non efficacia', alpha=0.7)

plt.yticks(range(len(df_noac)), df_noac['NOAC'], fontsize=12)
plt.xlabel('Hazard Ratio (IC 95%) vs Warfarin', fontsize=12)
plt.title('Efficacia dei NOAC nel ridurre lo stroke/embolia sistemica (vs Warfarin)', fontsize=14)
plt.grid(axis='x', linestyle=':', alpha=0.6)
plt.xlim(0.4, 1.2)
plt.legend(loc='lower right', fontsize=10)
plt.tight_layout()
plt.show()
```

### Differenze Cliniche: Rigetto Acuto vs. Tossicità da Inibitori della Calcineurina (CNI)

| Caratteristica | Rigetto Acuto (Cellulare o Umorale) | Tossicità da CNI (Ciclosporina/Tacrolimus) |
| :--- | :--- | :--- |
| **Tempistica** | Solitamente entro i primi 6 mesi (ma possibile sempre). | Correlata a livelli ematici elevati o uso cronico. |
| **Sintomi Sistemici** | Febbre, malessere, dolore/tensione sulla sede del trapianto. | Spesso asintomatica; possibile tremore, ipertensione, parestesie. |
| **Creatinina** | Aumento rapido e significativo. | Aumento graduale o fluttuante correlato al dosaggio. |
| **Diuresi** | Può ridursi (oliguria). | Solitamente conservata nelle fasi iniziali. |
| **Livelli Farmacologici** | Spesso bassi o nel range terapeutico. | Tipicamente elevati (sopra il range terapeutico). |
| **Sedimento Urinario** | Possibile proteinuria, ematuria, cilindri leucocitari. | Solitamente silente. |

### Analisi Bioptica (Standard d'Oro)

Quando il sospetto clinico non è dirimente, la biopsia renale secondo la **classificazione di Banff** è essenziale:

1.  **Rigetto Acuto Cellulare (TCMR):**
    *   **Tubulite:** Infiltrazione di linfociti nell'epitelio tubulare.
    *   **Infiammazione Interstiziale:** Edema e infiltrato linfocitario diffuso.
    *   **Endotelialite:** Linfociti sotto l'endotelio dei vasi arteriosi.

2.  **Tossicità da CNI:**
    *   **Forma Acuta:** Vasocostrizione dell'arteriola afferente (reversibile), vacuolizzazione isometrica delle cellule tubulari.
    *   **Forma Cronica:** Ialinosi arteriolare a 'scatto di dente', fibrosi interstiziale a strisce e atrofia tubulare.

### Gestione Clinica
*   **Sospetto Rigetto:** Richiede spesso boli di metilprednisolone o anticorpi antitimociti.
*   **Sospetto Tossicità:** Richiede la riduzione del dosaggio del CNI o lo shift verso altri farmaci (es. inibitori di mTOR come Sirolimus) sotto stretto monitoraggio della funzionalità renale.

### Analisi del Forest Plot sui NOAC

- **Intervallo di Confidenza (IC 95%)**: La barra orizzontale rappresenta l'intervallo di confidenza al 95%. Se l'intera barra si trova a sinistra della linea rossa verticale (HR=1.0), il risultato è statisticamente significativo a favore del NOAC.

- **Significatività Clinica**: Sebbene i valori specifici di HR possano variare leggermente tra i diversi NOAC, la classe nel suo complesso dimostra un profilo favorevole, con alcuni farmaci che mostrano una riduzione del rischio leggermente più marcata.

È importante notare che questi dati sono esemplificativi e le conclusioni cliniche dovrebbero sempre basarsi su metanalisi e linee guida aggiornate che considerano anche il profilo di sicurezza (es. rischio emorragico).

### Analisi del Forest Plot sul Profilo di Sicurezza dei NOAC (Rischio Emorragico Maggiore)

Questo Forest Plot illustra il rischio di sanguinamento maggiore associato ai Nuovi Anticoagulanti Orali (NOAC) rispetto al warfarin. Un Hazard Ratio (HR) inferiore a 1.0 indica un rischio emorragico minore rispetto al warfarin, mentre un HR superiore a 1.0 indica un rischio maggiore.

- **Hazard Ratio (HR)**: I valori di HR mostrano che alcuni NOAC tendono ad avere un rischio di sanguinamento maggiore inferiore o paragonabile a quello del warfarin.
- **Intervallo di Confidenza (IC 95%)**: Le barre orizzontali indicano l'intervallo di confidenza. Se l'intervallo attraversa la linea blu verticale (HR=1.0), la differenza nel rischio emorragico non è statisticamente significativa. Se l'intero intervallo è a sinistra della linea, il NOAC ha un rischio significativamente inferiore; se è a destra, il rischio è significativamente maggiore.
  - Dai dati esemplificativi, si può notare come alcuni NOAC possano presentare un rischio emorragico maggiore inferiore al warfarin (es. Apixaban), mentre altri possono essere comparabili o leggermente superiori (es. Rivaroxaban con un IC che sfiora o supera l'unità). Dabigatran mostra un IC che include l'unità, suggerendo una non-inferiorità o comparabilità al warfarin per il rischio emorragico generale.

**Considerazioni Cliniche**: È fondamentale bilanciare l'efficacia nella prevenzione dello stroke con il profilo di sicurezza emorragico. La scelta del NOAC dipende dalle caratteristiche del paziente, inclusi i fattori di rischio individuali per lo stroke e il sanguinamento. Le linee guida cliniche raccomandano una valutazione personalizzata per ottimizzare il rapporto beneficio-rischio.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Dati per l'efficacia (riduzione dello stroke)
data_efficacy = {
    'NOAC': ['Apixaban', 'Dabigatran', 'Edoxaban', 'Rivaroxaban'],
    'HR': [0.79, 0.65, 0.79, 0.82], # Hazard Ratio vs Warfarin per l'endpoint primario (stroke/embolia sistemica)
    'Lower_CI': [0.69, 0.52, 0.68, 0.70],
    'Upper_CI': [0.90, 0.81, 0.92, 0.96],
    'Type': 'Efficacia (Stroke)'
}
df_efficacy = pd.DataFrame(data_efficacy)

# Dati per la sicurezza (rischio emorragico maggiore)
data_bleeding = {
    'NOAC': ['Apixaban', 'Dabigatran', 'Edoxaban', 'Rivaroxaban'],
    'HR': [0.70, 0.90, 0.80, 0.95], # Hazard Ratio per sanguinamento maggiore vs Warfarin
    'Lower_CI': [0.60, 0.75, 0.70, 0.80],
    'Upper_CI': [0.82, 1.05, 0.92, 1.10],
    'Type': 'Sicurezza (Emorragia)'
}
df_bleeding = pd.DataFrame(data_bleeding)

# Unione dei due dataframe
df_combined = pd.concat([df_efficacy, df_bleeding]).reset_index(drop=True)

plt.figure(figsize=(12, 8))

# Creazione del Forest Plot combinato
# Utilizziamo colori diversi per distinguere efficacia e sicurezza
colors = {'Efficacia (Stroke)': 'darkgreen', 'Sicurezza (Emorragia)': 'darkred'}
markers = {'Efficacia (Stroke)': 'o', 'Sicurezza (Emorragia)': 's'}

# Per posizionare i punti vicini ma distinti per NOAC
offset = {'Efficacia (Stroke)': -0.1, 'Sicurezza (Emorragia)': 0.1}

for i, noac in enumerate(df_efficacy['NOAC']):
    # Efficacia
    eff_data = df_combined[(df_combined['NOAC'] == noac) & (df_combined['Type'] == 'Efficacia (Stroke)')].iloc[0]
    plt.errorbar(x=eff_data['HR'], y=i + offset['Efficacia (Stroke)'],
                 xerr=[[eff_data['HR'] - eff_data['Lower_CI']], [eff_data['Upper_CI'] - eff_data['HR']]],
                 fmt=markers['Efficacia (Stroke)'], color=colors['Efficacia (Stroke)'], capsize=8, markersize=10, zorder=3)
    plt.text(eff_data['Upper_CI'] + 0.02, i + offset['Efficacia (Stroke)'],
             f"HR = {eff_data['HR']:.2f} ({eff_data['Lower_CI']:.2f}-{eff_data['Upper_CI']:.2f})",
             va='center', ha='left', fontsize=9, color=colors['Efficacia (Stroke)'])

    # Sicurezza
    bleed_data = df_combined[(df_combined['NOAC'] == noac) & (df_combined['Type'] == 'Sicurezza (Emorragia)')].iloc[0]
    plt.errorbar(x=bleed_data['HR'], y=i + offset['Sicurezza (Emorragia)'],
                 xerr=[[bleed_data['HR'] - bleed_data['Lower_CI']], [bleed_data['Upper_CI'] - bleed_data['HR']]],
                 fmt=markers['Sicurezza (Emorragia)'], color=colors['Sicurezza (Emorragia)'], capsize=8, markersize=10, zorder=3)
    plt.text(bleed_data['Upper_CI'] + 0.02, i + offset['Sicurezza (Emorragia)'],
             f"HR = {bleed_data['HR']:.2f} ({bleed_data['Lower_CI']:.2f}-{bleed_data['Upper_CI']:.2f})",
             va='center', ha='left', fontsize=9, color=colors['Sicurezza (Emorragia)'])


plt.axvline(x=1.0, color='gray', linestyle='--', label='Linea di non differenza/efficacia', alpha=0.7)

plt.yticks(range(len(df_efficacy['NOAC'])), df_efficacy['NOAC'], fontsize=12)
plt.xlabel('Hazard Ratio (IC 95%) vs Warfarin', fontsize=12)
plt.title('Confronto Efficacia (Stroke) vs Sicurezza (Emorragia) dei NOAC vs Warfarin', fontsize=14)
plt.grid(axis='x', linestyle=':', alpha=0.6)
plt.xlim(0.4, 1.3) # Adattare il limite x per accogliere entrambi i tipi di dati

# Legenda manuale per tipi di dati (efficacia vs sicurezza)
handles = [
    plt.Line2D([0], [0], marker=markers['Efficacia (Stroke)'], color='w', label='Efficacia (Stroke)', markerfacecolor=colors['Efficacia (Stroke)'], markersize=10),
    plt.Line2D([0], [0], marker=markers['Sicurezza (Emorragia)'], color='w', label='Sicurezza (Emorragia)', markerfacecolor=colors['Sicurezza (Emorragia)'], markersize=10)
]
plt.legend(handles=handles, loc='lower right', fontsize=10)

plt.tight_layout()
plt.show()
```

### Analisi del Forest Plot Combinato: Efficacia (Stroke) vs Sicurezza (Rischio Emorragico) dei NOAC

Questo Forest Plot integrato presenta un confronto visivo diretto dell'efficacia (riduzione dello stroke) e della sicurezza (rischio di sanguinamento maggiore) per ciascun NOAC rispetto al warfarin. La linea verticale grigia a HR=1.0 indica l'assenza di differenza.

```python
def med_research_agent(query):
    system_prompt = """Sei un assistente IA specializzato in ricerca medica avanzata e farmacoterapia per medici specialisti.
    Il tuo obiettivo è fornire analisi basate sull'evidenza (EBM) con un focus rigoroso sui trial clinici.

    Regole Mandatorie:
    1. Cita esplicitamente i trial clinici cardine (es. EMPEROR, DAPA-CKD, LEADER) e l'anno di pubblicazione se disponibile.
    2. Usa terminologia medica di alto livello (es. farmacocinetica, endpoint compositi, HR, IC 95%).
    3. Struttura: Meccanismo Molecolare, Analisi Trial Clinici (Evidenze), Considerazioni Cliniche e Linee Guida (ESC/EASD/KDIGO).
    4. Se i dati sono contrastanti tra diversi trial, evidenzia la discrepanza.
    5. Mantieni un tono professionale, sintetico e analitico."""

    response = llm.create_chat_completion(
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": query}
        ],
        max_tokens=2048,
        temperature=0.2,
        top_p=0.95,
        repeat_penalty=1.2
    )
    return response['choices'][0]['message']['content']

print("System Prompt ottimizzato per il richiamo dei trial clinici.")
```
