## Filter and Format Questions

### Subtask:
Filter the dataset to include only questions that start with 'Does the gene ' and apply the `format_example` function to each of these questions. The `format_example` function adds a preamble with instructions for the agent and formats the question into a prompt suitable for the model.


**Reasoning**:
I will define the `format_example` function as specified, which will create a prompt with an instructional preamble and a target answer from the original question and answer fields. This function is a prerequisite for formatting the questions.



```python
def format_example(example):
    prompt = f"Answer the following question based on the provided context or your general knowledge. If the question is about a gene interaction, describe the interaction. If it's about a gene's function, describe its function. If it's about a gene's presence or absence, state that. \n\nQuestion: {example['question']}"
    target = example['answer']
    return {'prompt': prompt, 'target': target}
```

**Reasoning**:
Now that the `format_example` function is defined, I will filter the 'train' split of the loaded `dataset` to select only those examples where the 'question' field begins with the string 'Does the gene '. Then, I will apply the `format_example` function to this filtered dataset using the `.map()` method to create new 'prompt' and 'target' columns, as per the subtask instructions.



```python
gene_questions = dataset['train'].filter(lambda example: example['question'].startswith('Does the gene '))
formatted_dataset = gene_questions.map(format_example)

print("Filtered and formatted dataset:")
print(formatted_dataset)
print("\nFirst 5 formatted prompts:")
for i in range(5):
    print(f"Prompt {i+1}: {formatted_dataset[i]['prompt']}")
```

## Sviluppo della Logica di Ricerca e Filtraggio

### Subtask:
Implementare una funzione Python che prenda una query medica e simuli la 'ricerca' attraverso i dati simulati, applicando una logica di filtraggio di base per identificare le informazioni più pertinenti.


**Reasoning**:
I will modify the `simulate_medical_data_access` function to accumulate all matching results from the `medical_data` list instead of returning the first one. The function will return a list of results or a default message if no matches are found, and then I will test it with example queries.



```python
def simulate_medical_data_access(query):
    query_lower = query.lower()
    matching_results = []
    for entry in medical_data:
        for keyword in entry['query_keywords']:
            if keyword.lower() in query_lower:
                matching_results.append(entry['result'])
                break # Move to the next entry once a keyword is matched for the current entry

    if not matching_results:
        return "No specific information found for this query in the simulated database."
    else:
        return matching_results

print("Modified simulated medical data access function defined.")

# Test the modified function with example queries
print("\nTesting modified simulate_medical_data_access function:")
print(f"Query: 'What are the interactions between Warfarin and Aspirin?'\nResult: {simulate_medical_data_access('What are the interactions between Warfarin and Aspirin?')}")
print(f"Query: 'How effective is Metformin for diabetes?'\nResult: {simulate_medical_data_access('How effective is Metformin for diabetes?')}")
print(f"Query: 'What is the prognosis for glioblastoma?'\nResult: {simulate_medical_data_access('What is the prognosis for glioblastoma?')}")
print(f"Query: 'Tell me about side effects of NSAIDs like Ibuprofen?'\nResult: {simulate_medical_data_access('Tell me about side effects of NSAIDs like Ibuprofen?')}")
print(f"Query: 'How does immunotherapy fight cancer?'\nResult: {simulate_medical_data_access('How does immunotherapy fight cancer?')}")
print(f"Query: 'Tell me about metformin, side effects and efficacy.'\nResult: {simulate_medical_data_access('Tell me about metformin, side effects and efficacy.')}") # Multiple matches
print(f"Query: 'What is the best treatment for headache?'\nResult: {simulate_medical_data_access('What is the best treatment for headache?')}") # No specific match
```

## Formattazione Dettagliata dei Risultati

### Subtask:
Sviluppare una funzione per formattare i risultati della ricerca in un rapporto dettagliato e leggibile, adatto a un medico. Questo includerà la strutturazione delle informazioni in modo chiaro e conciso.


**Reasoning**:
I will define the `format_medical_report` function as specified, which will take a query and a list of results, and format them into a readable medical report. This function will include logic to handle cases with single, multiple, or no results. After defining the function, I will test it with various scenarios to ensure it meets the requirements.



```python
def format_medical_report(query, results):
    report = f"# Rapporto di Ricerca Medica\n\n"
    report += f"## Query Originale: {query}\n\n"

    if not results or (isinstance(results, list) and len(results) == 1 and "No specific information found" in results[0]):
        report += "### Nessuna Informazione Specificata Trovata\n\n"
        report += "Non è stata trovata alcuna informazione specifica per la query nella banca dati simulata. Si prega di provare una query diversa o di consultare fonti aggiuntive.\n"
    else:
        report += "### Risultati della Ricerca:\n\n"
        if isinstance(results, list):
            for i, res in enumerate(results):
                report += f"**{i+1}.** {res}\n\n"
        else:
            report += f"**1.** {results}\n\n"
    return report.strip()

print("Medical report formatting function defined.")

# Test the format_medical_report function
print("\n--- Testing format_medical_report function ---")

# Test Case 1: Single result
query1 = 'What are the interactions between Warfarin and Aspirin?'
results1 = simulate_medical_data_access(query1)
print(f"\nReport for query: '{query1}'")
print(format_medical_report(query1, results1))

# Test Case 2: Multiple results (using a query that should trigger multiple matches if available, otherwise it will match one and return it as a list)
query2 = 'Tell me about metformin, side effects and efficacy.'
results2 = simulate_medical_data_access(query2)
print(f"\nReport for query: '{query2}'")
print(format_medical_report(query2, results2))

# Test Case 3: No specific information found
query3 = 'What is the best treatment for common cold?'
results3 = simulate_medical_data_access(query3)
print(f"\nReport for query: '{query3}'")
print(format_medical_report(query3, results3))

# Test Case 4: Another single result
query4 = 'What is the prognosis for glioblastoma?'
results4 = simulate_medical_data_access(query4)
print(f"\nReport for query: '{query4}'")
print(format_medical_report(query4, results4))
```
