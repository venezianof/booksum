### Integrazione Database Esterni
Per rendere l'agente operativo su dati reali, possiamo utilizzare librerie come `Biopython` per interrogare **PubMed** o fare richieste REST a **ClinicalTrials.gov**. Di seguito è mostrato come integrare PubMed come `Tool` di LangChain.

```python
from Bio import Entrez

# Configura la tua email per PubMed
Entrez.email = "your-email@example.com"

def pubmed_search(query):
    """Esegue una ricerca reale su PubMed e restituisce gli ID degli articoli."""
    handle = Entrez.esearch(db="pubmed", term=query, retmax=5)
    record = Entrez.read(handle)
    handle.close()
    ids = record["IdList"]
    return f"Trovati {len(ids)} articoli su PubMed. IDs: {', '.join(ids)}"

# Aggiornamento degli strumenti dell'agente
tools.append(
    Tool(
        name="PubMed_Search",
        func=pubmed_search,
        description="Cerca articoli scientifici reali nel database PubMed (NIH)."
    )
)

print("Strumento PubMed integrato con successo.")
```

### Inizializzazione dell'Agente con Gemini e Strumenti Aggiornati

Ora, inizializzeremo l'agente con il modello Gemini e tutti gli strumenti disponibili (inclusa la ricerca web e PubMed). Assicurati che la tua `GOOGLE_API_KEY` sia impostata nei segreti di Colab come indicato in precedenza nel notebook.

```python
import os
import langchain
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata
from langchain_core.prompts import PromptTemplate

# 1. Check for API Key
try:
    GOOGLE_API_KEY = userdata.get("GOOGLE_API_KEY")
except Exception:
    GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY")

if not GOOGLE_API_KEY:
    print("❌ ERRORE: GOOGLE_API_KEY non trovata.")
    print("Per favore, aggiungi la tua API Key nei 'Secrets' (icona 🔑 a sinistra) con il nome GOOGLE_API_KEY.")
else:
    # 2. Setup Agent Components
    try:
        from langchain.agents import AgentExecutor, create_react_agent
    except ImportError:
        from langchain.agents.agent import AgentExecutor
        from langchain.agents.react.agent import create_react_agent

    print(f"LangChain version: {langchain.__version__}")

    llm = ChatGoogleGenerativeAI(model='gemini-1.5-pro', google_api_key=GOOGLE_API_KEY)

    if 'tools' not in locals() or not tools:
        from langchain_core.tools import Tool
        def placeholder_func(query): return 'Search not available'
        tools = [Tool(name='Search', func=placeholder_func, description='Placeholder')]

    template = """Answer the following questions as best you can. You have access to the following tools:

{tools}

Use the following format:

Question: the input question you must answer
Thought: you should always think about what to do
Action: the action to take, should be one of [{tool_names}]
Action Input: the input to the action
Observation: the result of the action
... (this Thought/Action/Action Input/Observation can repeat N times)
Thought: I now know the final answer
Final Answer: the final answer to the original input question

Begin!

Question: {input}
Thought:{agent_scratchpad}"""

    prompt = PromptTemplate.from_template(template)

    try:
        agent = create_react_agent(llm, tools, prompt)
        agent_executor = AgentExecutor(
            agent=agent,
            tools=tools,
            verbose=True,
            handle_parsing_errors=True,
            max_iterations=5
        )
        print("✅ Agente pronto per la ricerca.")
    except Exception as e:
        print(f"❌ Errore durante l'inizializzazione: {e}")
```

### Riassunto Linee Guida ESC 2024/2025 sulla Fibrillazione Atriale

Le nuove linee guida della European Society of Cardiology (ESC) propongono una visione più olistica e dinamica.

#### 1. Il nuovo schema CARE
Sostituisce il vecchio schema ABC:
*   **C** (Comorbidity & risk factor management): Trattamento dei fattori di rischio.
*   **A** (Avoid stroke): Prevenzione dell'ictus tramite anticoagulazione.
*   **R** (Reduced symptoms via rate & rhythm control): Gestione dei sintomi.
*   **E** (Evaluation & reassessment): Rivalutazione continua del paziente.

#### 2. Focus sull'Ablazione
L'ablazione della fibrillazione atriale riceve una raccomandazione di **Classe I** in molti scenari come terapia iniziale per migliorare la qualità della vita e prevenire la progressione della malattia.

#### 3. Screening con Wearables
L'ESC riconosce il ruolo dei sensori digitali, ma sottolinea la necessità di conferma medica per evitare sovradiagnosi e ansia nel paziente.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 1. Grafico a barre per lo schema CARE
care_data = {
    'Pilastro': ['C - Comorbidità', 'A - Anticoagulazione', 'R - Controllo Ritmo', 'E - Rivalutazione'],
    'Focus': [95, 100, 90, 85]
}
df_care = pd.DataFrame(care_data)

plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='Focus', y='Pilastro', data=df_care, palette='viridis', hue='Pilastro', legend=False)
plt.title('Priorità Linee Guida ESC 2025: Schema CARE', fontsize=15)
plt.xlabel('Grado di enfasi clinica (Simulato)', fontsize=12)
plt.xlim(0, 110)
plt.show()
```

```python
# 2. Simulazione del successo nel mantenimento del ritmo sinusale
time = np.linspace(0, 24, 100) # 24 mesi
ablazione = 100 * np.exp(-0.01 * time) + np.random.normal(0, 2, 100)
farmaci = 100 * np.exp(-0.04 * time) + np.random.normal(0, 2, 100)

plt.figure(figsize=(12, 6))
plt.plot(time, ablazione, label='Ablazione Precoce (Classe I)', color='blue', linewidth=2.5)
plt.plot(time, farmaci, label='Sola Terapia Farmacologica', color='red', linestyle='--', linewidth=2)

plt.title('Mantenimento del Ritmo Sinusale a 24 mesi (Simulazione Trend ESC 2025)', fontsize=14)
plt.xlabel('Mesi dal Trattamento', fontsize=12)
plt.ylabel('% Pazienti in Ritmo Sinusale', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

### Gestione dei Fattori di Rischio (Pilastro 'C' dello schema CARE)

Le linee guida ESC 2025 pongono un'enfasi senza precedenti sulla modifica dei fattori di rischio come pre-requisito per il successo di qualsiasi terapia del ritmo.

#### 1. Obesità e Calo Ponderale
*   **Target**: Indice di Massa Corporea (BMI) < 27 kg/m².
*   **Evidenza**: Un calo ponderale sostenuto (>10% del peso corporeo) è associato a una riduzione significativa del burden di AF e può persino indurre la regressione da forme persistenti a parossistiche.
*   **Novità**: Considerare l'uso di agonisti del recettore GLP-1 in pazienti selezionati per favorire il raggiungimento dei target metabolici.

#### 2. Ipertensione Arteriosa
*   **Target**: Pressione < 130/80 mmHg.
*   **Meccanismo**: L'ipertensione causa rimodellamento atriale e fibrosi; il controllo rigoroso riduce la probabilità di recidiva dopo ablazione.

#### 3. Altri Fattori
*   **Apnee Notturne (OSA)**: Screening obbligatorio e trattamento con CPAP.
*   **Alcol**: Raccomandata l'astinenza o la riduzione drastica (limite < 3 unità/settimana).

```python
import matplotlib.pyplot as plt

# Visualizzazione dell'impatto del calo ponderale sulla libertà da AF (Dati simulati basati su studi ARREST-AF/LEGACY)
gruppi = ['Calo <3%', 'Calo 3-9%', 'Calo >10%']
liberta_da_af = [35, 50, 85] # % di pazienti liberi da aritmia a 5 anni

plt.figure(figsize=(10, 5))
plt.bar(gruppi, liberta_da_af, color=['#ff9999','#66b3ff','#99ff99'])
plt.title('Impatto del Calo Ponderale sulla Libertà da Fibrillazione Atriale', fontsize=14)
plt.ylabel('% Pazienti Liberi da AF (Follow-up lungo termine)', fontsize=12)
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for i, v in enumerate(liberta_da_af):
    plt.text(i, v + 2, f"{v}%", ha='center', fontweight='bold')

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Dati simulati basati sull'entità dell'effetto riportata in letteratura per le linee guida ESC
impact_data = {
    'Fattore di Rischio': [
        'Calo Ponderale >10%',
        'Controllo BP (<130/80)',
        'Trattamento CPAP (OSA)',
        'Astinenza Alcol',
        'Esercizio Fisico Moderato'
    ],
    'Riduzione Recidive (%)': [45, 38, 25, 20, 30]
}

df_impact = pd.DataFrame(impact_data).sort_values(by='Riduzione Recidive (%)', ascending=True)

plt.figure(figsize=(12, 7))
colors = plt.cm.RdYlGn(df_impact['Riduzione Recidive (%)'] / 50)

plt.barh(df_impact['Fattore di Rischio'], df_impact['Riduzione Recidive (%)'], color=colors)
plt.title('Impatto Relativo della Gestione dei Fattori di Rischio sulla Riduzione delle Recidive AF', fontsize=14)
plt.xlabel('Efficacia Stimata nella Riduzione delle Recidive (%)', fontsize=12)
plt.xlim(0, 60)

for i, v in enumerate(df_impact['Riduzione Recidive (%)']):
    plt.text(v + 1, i, f'+{v}%', va='center', fontweight='bold')

plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

### 📋 Riepilogo Azioni Prioritarie: Lo Schema CARE (ESC 2025)

Le linee guida ESC 2025 hanno consolidato la gestione della AF in quattro pilastri fondamentali per garantire un approccio olistico:

#### **1. C (Comorbidity & Risk Factor Management)**
*   **Azione**: Identificare e trattare aggressivamente i fattori di rischio modificabili.
*   **Priorità**: Calo ponderale (target BMI < 27 kg/m²), controllo pressorio (<130/80 mmHg), gestione del diabete (GLP-1 RA se indicato) e screening per le apnee notturne (OSA).

#### **2. A (Avoid Stroke)**
*   **Azione**: Prevenzione tromboembolica basata sul rischio individuale.
*   **Priorità**: Utilizzo dei NOAC (anticoagulanti orali non antagonisti della vitamina K) come prima linea. Rivalutazione periodica del rischio emorragico per ottimizzare la terapia anziché sospenderla.

#### **3. R (Reduced Symptoms via Rate & Rhythm Control)**
*   **Azione**: Migliorare la qualità della vita attraverso il controllo della frequenza o del ritmo.
*   **Priorità**: Raccomandazione di **Classe I** per l'ablazione transcatetere precoce nei pazienti con AF parossistica/persistente e in quelli con insufficienza cardiaca (HFrEF).

#### **4. E (Evaluation & Reassessment)**
*   **Azione**: Monitoraggio continuo e adattamento della terapia nel tempo.
*   **Priorità**: Rivalutazione clinica periodica per intercettare la progressione della AF, nuovi fattori di rischio o cambiamenti nel profilo di rischio tromboembolico/emorragico.

### Tabella di Confronto: Terapia Farmacologica vs Ablazione Transcatetere (ESC 2025)

| Caratteristica | Terapia Farmacologica (Antiaritmici) | Ablazione Transcatetere |
| :--- | :--- | :--- |
| **Efficacia (Mantenimento Ritmo)** | Moderata (spesso richiede più tentativi) | Elevata (superiore ai farmaci nel lungo termine) |
| **Raccomandazione ESC** | Classe I (scelta iniziale in molti pazienti) | Classe I (raccomandata precocemente, specie in HF) |
| **Profilo di Sicurezza** | Effetti collaterali extracardiaci e pro-aritmia | Rischi procedurali (tamponamento, fistola) < 1% |
| **Qualità della Vita** | Miglioramento variabile | Miglioramento significativo e sostenuto |
| **Progressione AF** | Non previene sempre la progressione a permanente | Riduce significativamente il burden di AF |
| **Gestione** | Richiede aderenza quotidiana a lungo termine | Procedura singola (possibile re-do) |
