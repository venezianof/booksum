### 5. Medical Research & Therapy Agent Setup
This section initializes the specialized agent for medical inquiries and therapy development.

```python
!pip install biopython transformers langchain-community google-search-results langchain-google-genai langchain
```

```python
import os
from langchain.agents import initialize_agent, Tool
from langchain.agents import AgentType
from langchain_google_genai import ChatGoogleGenerativeAI

# Configure the Medical Research Agent
def medical_research_logic(query):
    """Simulates deep research in medical databases."""
    return f"Risultati della ricerca profonda per: {query}. Analisi dei trial clinici e della letteratura peer-reviewed in corso..."

tools = [
    Tool(
        name="Medical_Database",
        func=medical_research_logic,
        description="Utilizzare per ricerche mediche approfondite e letteratura scientifica."
    )
]

# Initialize with Gemini
# llm = ChatGoogleGenerativeAI(model="gemini-1.5-pro", google_api_key="YOUR_API_KEY")
# agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True)

print("Agente di ricerca medica inizializzato. Pronto per l'analisi terapeutica.")
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

def therapy_generator(condition):
    """Generates a structured therapeutic protocol."""
    return f"Protocollo suggerito per {condition}: 1. Analisi biomarcatori 2. Selezione farmaci di precisione 3. Monitoraggio trial fase III."

tools.append(
    Tool(
        name="Therapy_Protocol_Generator",
        func=therapy_generator,
        description="Genera protocolli terapeutici strutturati per una specifica condizione medica."
    )
)

print("Strumenti di terapia aggiunti. L'agente è ora in grado di generare protocolli.")
```

### Integrazione Database Esterni
Per rendere l'agente operativo su dati reali, possiamo utilizzare librerie come `Biopython` per interrogare **PubMed** o fare richieste REST a **ClinicalTrials.gov**. Di seguito è mostrato come integrare PubMed come `Tool` di LangChain.

```python
from Bio import Entrez

# Configura la tua email per PubMed
Entrez.email = "your-email@example.com"

def pubmed_search(query):
    """Esegue una ricerca reale su PubMed e restituisce gli ID degli articoli."""
    handle = Entrez.esearch(db="pubmed", term=query, retmax=5)
    record = Entrez.read(handle)
    handle.close()
    ids = record["IdList"]
    return f"Trovati {len(ids)} articoli su PubMed. IDs: {', '.join(ids)}"

# Aggiornamento degli strumenti dell'agente
tools.append(
    Tool(
        name="PubMed_Search",
        func=pubmed_search,
        description="Cerca articoli scientifici reali nel database PubMed (NIH)."
    )
)

print("Strumento PubMed integrato con successo.")
```

### Configurazione Chiave API SerpAPI

Per utilizzare lo strumento di ricerca web (`GoogleSearchAPIWrapper`), avrai bisogno di una chiave API SerpAPI. Se non ne hai una, puoi ottenerla registrandoti su [SerpAPI](https://serpapi.com/).

Una volta ottenuta la chiave, aggiungila ai segreti di Colab (`🔑` nel pannello a sinistra) con il nome `SERPAPI_API_KEY`.

```python
from langchain_community.utilities import GoogleSearchAPIWrapper
from langchain.agents import AgentType, initialize_agent, Tool
from langchain_google_genai import ChatGoogleGenerativeAI
import os

# Carica la chiave API di SerpAPI dai segreti di Colab
SERPAPI_API_KEY = os.getenv("SERPAPI_API_KEY") # For Colab secrets, usually set as environment variables

if not SERPAPI_API_KEY:
    print("ERRORE: La chiave API SERPAPI_API_KEY non è stata trovata nei segreti di Colab. Assicurati di averla impostata correttamente.")
else:
    # Inizializza la Google Search API
    search = GoogleSearchAPIWrapper(serpapi_api_key=SERPAPI_API_KEY)

    # Aggiungi lo strumento di ricerca web agli strumenti dell'agente
    tools.append(
        Tool(
            name="Google_Search",
            func=search.run,
            description="Utilizzare per rispondere a domande generiche basate su conoscenze attuali."
        )
    )
    print("Strumento Google Search integrato con successo.")
```

### Inizializzazione dell'Agente con Gemini e Strumenti Aggiornati

Ora, inizializzeremo l'agente con il modello Gemini e tutti gli strumenti disponibili (inclusa la ricerca web e PubMed). Assicurati che la tua `GOOGLE_API_KEY` sia impostata nei segreti di Colab come indicato in precedenza nel notebook.

```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.agents import create_react_agent
from langchain.agents.agent_executor import AgentExecutor
from langchain_core.prompts import ChatPromptTemplate

# Assicurati che la GOOGLE_API_KEY sia impostata nei segreti di Colab
GOOGLE_API_KEY = os.getenv("GOOGLE_API_KEY") # Assuming it's set as environment variable from Colab secrets

if not GOOGLE_API_KEY:
    print("ERRORE: La chiave API GOOGLE_API_KEY non è stata trovata nei segreti di Colab. Assicurati di averla impostata correttamente.")
else:
    llm = ChatGoogleGenerativeAI(model="gemini-1.5-pro", google_api_key=GOOGLE_API_KEY)

    # Define the prompt for the ReAct agent
    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are a helpful assistant."),
        ("human", "{input}"),
        ("placeholder", "{agent_scratchpad}"),
    ])

    # Create the ReAct agent
    agent = create_react_agent(llm, tools, prompt)

    # Create the AgentExecutor
    agent_executor = AgentExecutor(
        agent=agent,
        tools=tools,
        verbose=True,
        handle_parsing_errors=True # Aggiungi questo per gestire meglio gli errori di parsing
    )
    print("Agente LangChain inizializzato con Gemini e strumenti aggiornati.")
```

### Esecuzione della Ricerca per le Linee Guida ESC 2025 sulla Fibrillazione Atriale

Ora che l'agente è configurato, possiamo chiedergli di cercare le linee guida.

```python
if 'agent_executor' in locals() and agent_executor is not None:
    query = "ESC 2025 atrial fibrillation guidelines"
    print(f"Esecuzione della query: {query}")
    response = agent_executor.invoke({"input": query})
    print(response)
else:
    print("L'agente non è stato inizializzato correttamente. Verifica le chiavi API.")
```
